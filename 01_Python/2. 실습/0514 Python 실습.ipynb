````python
import re

def regex_examples():
    # ë¬¸ì í´ë˜ìŠ¤ [] ì˜ˆì œ
    print(re.search(r"[a]", "a"))      # a í¬í•¨ â†’ match
    print(re.search(r"[a]", "ba"))     # a í¬í•¨ â†’ match
    print(re.search(r"[a]", "abb"))    # a í¬í•¨ â†’ match
    print(re.search(r"[a]", "cb"))     # a ì—†ìŒ â†’ no match

    print("-----------------------------------")
    print(re.search(r"[abc]", "apple"))     # a â†’ match
    print(re.search(r"[abc]", "dog"))       # a,b,c ì—†ìŒ â†’ no match
    print(re.search(r"[0-9]", "2024year"))  # 2 â†’ match
    print(re.search(r"[0-9]", "year"))      # ìˆ«ì ì—†ìŒ â†’ no match
    print(re.search(r"[A-Z]", "Hello"))     # H â†’ match
    print(re.search(r"[A-Z]", "hello"))     # ëŒ€ë¬¸ì ì—†ìŒ â†’ no match
    print(re.search(r"[a-z]", "Hello"))     # e â†’ match
    print(re.search(r"[A-Za-z]", "123Abc")) # A â†’ match

    print("-----------------------------------")
    print(re.search(r"[^a]", "aaa"))    # ëª¨ë‘ a â†’ no match
    print(re.search(r"[^a]", "aab"))    # b â†’ match
    print(re.search(r"[^0-9]", "1234a"))# a â†’ match

    print(re.search(r"[a-z]{3}", "abc"))    # abc â†’ match
    print(re.search(r"[a-z]{3}", "a1c"))    # 1 ë•Œë¬¸ì— ì‹¤íŒ¨
    print(re.search(r"[A-Za-z]{2,5}", "Hi"))# Hi â†’ match

    print(re.search(r"[0-9][a-z]", "3b"))     # 3b â†’ match
    print(re.search(r"[0-9][a-z]", "a3b"))    # 3bì—ì„œ ë§¤ì¹˜ë¨
    print(re.search(r"[0-9][a-z]", "ab"))     # ìˆ«ìê°€ í¬í•¨ë˜ì–´ì•¼ â†’ no match

    print("===========================")
    print(re.search(r"^abc", "abcde"))   # abcë¡œ ì‹œì‘ â†’ match
    print(re.search(r"^abc", "abc")) 
    print(re.search(r"^abc", "abc12"))
    print(re.search(r"^abc",  "aabc12")) # aabcë¡œ ì‹œì‘ â†’ no match

    print(re.search(r"abc$",  "abc"))    # abcë¡œ ë â†’ match
    print(re.search(r"abc$",  "12abc"))  # abcë¡œ ë â†’ match
    print(re.search(r"abc$",  "ea abc")) # abcë¡œ ë â†’ match

    print("===========================")
    print(re.search(r"[p|P]ython",  "python"))  # p â†’ match
    print(re.search(r"[p|P]ython",  "Python"))  # P â†’ match
    print(re.search(r"[p|P]ython",  "PYTHON"))  # ëŒ€ì†Œë¬¸ì ëª¨ë‘ ëŒ€ë¬¸ì â†’ no match

    print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
    print(re.search(r"[^abc]",  "a"))      # abc ì œì™¸í•œ ë¬¸ìê°€ ì—†ìœ¼ë¯€ë¡œ no match
    print(re.search(r"[^abc]",  "b"))
    print(re.search(r"[^abc]",  "c"))
    print(re.search(r"[^abc]",  "aajdb"))  # j â†’ match
    print(re.search(r"[^abc]",  "dfc"))    # d â†’ match

    # ìˆ«ì 0ê°œ ë˜ëŠ” 1ê°œ
    print(re.fullmatch(r"\d?", ""))     # '' (ìˆ«ì ì—†ìŒ) â†’ match
    print(re.fullmatch(r"\d?", "5"))    # '5' (ìˆ«ì 1ê°œ) â†’ match
    print(re.fullmatch(r"\d?", "57"))   # '57' (2ê°œ) â†’ no match

    # ìˆ«ì í•œê°œ ì´ìƒ
    print(re.fullmatch(r"\d+", "5"))     # '5' â†’ match
    print(re.fullmatch(r"\d+", "123"))   # '123' â†’ match
    print(re.fullmatch(r"\d+", ""))      # ì—†ìŒ â†’ no match

    # ìˆ«ì 0ê°œ ì´ìƒ
    print(re.fullmatch(r"\d*", ""))       # '' (0ê°œ í—ˆìš©) â†’ match
    print(re.fullmatch(r"\d*", "12345"))  # '12345' â†’ match
    print(re.fullmatch(r"\d*", "a"))      # 'a'ëŠ” ìˆ«ìê°€ ì•„ë‹˜ â†’ no match

    # ì •í™•íˆ 3ìë¦¬ ìˆ«ì
    pattern = r"\d{3}"
    print(re.fullmatch(pattern, "123"))    # 3ìë¦¬ â†’ match
    print(re.fullmatch(pattern, "45"))     # 2ìë¦¬ â†’ no match
    print(re.fullmatch(pattern, "7890"))   # 4ìë¦¬ â†’ no match
    print(re.fullmatch(pattern, "abc"))    # ìˆ«ì ì•„ë‹˜ â†’ no match

    # 3~4ìë¦¬ ìˆ«ì
    pattern = r"\d{3,4}"
    print(re.fullmatch(pattern, "123"))     # 3ìë¦¬ â†’ match
    print(re.fullmatch(pattern, "1234"))    # 4ìë¦¬ â†’ match
    print(re.fullmatch(pattern, "12"))      # 2ìë¦¬ â†’ no match
    print(re.fullmatch(pattern, "12345"))   # 5ìë¦¬ â†’ no match

    # \b ë‹¨ì–´ ê²½ê³„ ì˜ˆì œ
    text = "I like apple and pineapple."
    match = re.search(r"\bapple\b", text)  # 'apple'ë§Œ ì°¾ê¸°
    print(match.group())  # apple

regex_examples()
````

````python
import re
# ? - ì—†ê±°ë‚˜ í•˜ë‚˜ë§Œ ìˆê±°ë‚˜ 
# + - íŒ¨í„´ì´ í•˜ë‚˜ì´ìƒ ë°˜ë³µ
# * - ì—†ê±°ë‚˜ í•˜ë‚˜ì´ìƒ ë°˜ë³µ 

patterns=[r"\d?", r"\d+", r"\d*"]
text = ["abc", "1abc", "12abc", "123", "aa12ab"]

for pattern in patterns:
    resultList=[]
    for item in text:
        result = re.search(pattern, item)
        if result == None:
            resultList.append(item+"-X")
        else:
            resultList.append(item+"-O")
print(resultList)
````
