### ğŸ”¤ % í¬ë§¤íŒ… í˜•ì‹ ì§€ì •ì

### ğŸ¯ ìë¦¿ìˆ˜ ì§€ì •
- `%10s`: 10ìë¦¬ í™•ë³´ í›„ ì˜¤ë¥¸ìª½ ì •ë ¬
- `%-10s`: 10ìë¦¬ í™•ë³´ í›„ ì™¼ìª½ ì •ë ¬

````python
# ë¬¸ìì—´ í¬ë§¤íŒ… ì™„ì „ ì •ë³µ
name = "í™ê¸¸ë™"
age = 34
height = 183.5
score = 89.7

print("=== 1. % í¬ë§¤íŒ… (ì „í†µì  ë°©ë²•) ===")
# ê¸°ë³¸ ì‚¬ìš©ë²•
result1 = "%sì˜ ë‚˜ì´ëŠ” %dì…ë‹ˆë‹¤. í‚¤ëŠ” %.1fì…ë‹ˆë‹¤." % (name, age, height)
print(result1)

# ì§„ë²• ë³€í™˜
number = 37
print(f"10ì§„ìˆ˜: {number}")
print("8ì§„ìˆ˜: %o" % number)
print("16ì§„ìˆ˜: %x%%" % number)  # %% = % ì¶œë ¥
print()

print("=== 2. ìë¦¿ìˆ˜ ì§€ì • ===")
# ì˜¤ë¥¸ìª½ ì •ë ¬ (ê¸°ë³¸)
print("'%10s' '%10s'" % ("hi", "hello"))
# ì™¼ìª½ ì •ë ¬
print("'%-10s' '%-10s'" % ("hi", "hello"))

# ì‹¤ìˆ˜ ìë¦¿ìˆ˜ ì§€ì •
pi = 3.141592653589793
print("ê¸°ë³¸: %f" % pi)
print("ì†Œìˆ˜ì  2ìë¦¬: %.2f" % pi)
print("ì „ì²´ 7ìë¦¬, ì†Œìˆ˜ì  2ìë¦¬: %7.2f" % pi)
print()

print("=== 3. .format() ë©”ì„œë“œ ===")
# ìœ„ì¹˜ ì¸ë±ìŠ¤ ì‚¬ìš©
text1 = "ì´ë¦„: {0}, ë‚˜ì´: {1}, {0}ì€ {1}ì‚´ì…ë‹ˆë‹¤.".format(name, age)
print(text1)

# ì¸ë±ìŠ¤ ìƒëµ (ìˆœì„œëŒ€ë¡œ)
text2 = "ì´ë¦„: {}, ë‚˜ì´: {}".format(name, age)
print(text2)

# í‚¤ì›Œë“œ ì¸ìˆ˜ ì‚¬ìš©
text3 = "ì´ë¦„: {student_name}, ë‚˜ì´: {student_age}".format(
    student_name=name, student_age=age
)
print(text3)

# ì •ë ¬ê³¼ íŒ¨ë”©
print("'{0:<10}' '{1:<10}'".format("hi", "hello"))  # ì™¼ìª½ ì •ë ¬
print("'{0:>10}' '{1:>10}'".format("hi", "hello"))  # ì˜¤ë¥¸ìª½ ì •ë ¬
print("'{0:^10}' '{1:^10}'".format("hi", "hello"))  # ê°€ìš´ë° ì •ë ¬
print("'{0:=^10}' '{1:*^10}'".format("hi", "hello"))  # ë¬¸ìë¡œ ì±„ìš°ê¸°

# ì‹¤ìˆ˜ í¬ë§·
print("{0:.4f}".format(pi))
print()

print("=== 4. f-string (Python 3.6+, ê¶Œì¥!) ===")
# ê°€ì¥ ê°„ë‹¨í•˜ê³  ì§ê´€ì 
print(f"ì´ë¦„: {name}, ë‚˜ì´: {age}")
print(f"í‚¤: {height}cm, ì„±ì : {score}ì ")

# í‘œí˜„ì‹ë„ ê°€ëŠ¥
print(f"{name}ì˜ BMI: {65 / ((height/100) ** 2):.1f}")

# ì‚¬ìš©ì ì…ë ¥ê³¼ í•¨ê»˜
x = int(input("x = "))
y = int(input("y = "))
print(f"{x} + {y} = {x + y}")
print(f"{x} Ã— {y} = {x * y}")

print("\nğŸ’¡ ê¶Œì¥ì‚¬í•­: Python 3.6 ì´ìƒì—ì„œëŠ” f-stringì„ ì‚¬ìš©í•˜ì„¸ìš”!")
````
