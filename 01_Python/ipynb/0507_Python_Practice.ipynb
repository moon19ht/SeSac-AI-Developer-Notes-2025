# í•¨ìˆ˜ê°€ ì‹¤í–‰ ì‹œê°„ì— ê²°ì •ë˜ëŠ” ë™ì  ê²°ì •(Dynamic Dispatch) ì˜ˆì‹œ

ì œë„ˆë ˆì´í„°(generator): ê°’ì„ í•˜ë‚˜ì”© ìƒì„±í•´ì„œ ìˆœíšŒí•  ìˆ˜ ìˆëŠ” í•¨ìˆ˜ë‚˜ ê°ì²´<br>

````python
# í•¨ìˆ˜ì—ì„œ ê°’ì„ ë°˜í™˜í•  ë•Œ returnê³¼ yieldì˜ ì°¨ì´
# return: ê°’ì„ ë°˜í™˜í•˜ê³  í•¨ìˆ˜ ì¢…ë£Œ
# yield: ê°’ì„ ë°˜í™˜í•˜ì§€ë§Œ í•¨ìˆ˜ëŠ” ì¢…ë£Œí•˜ì§€ ì•Šê³  ëŒ€ê¸° ìƒíƒœê°€ ë¨ (ì œë„ˆë ˆì´í„° í•¨ìˆ˜)

def myrange(start=1, end=5):
    """
    startë¶€í„° endê¹Œì§€ 1ì”© ì¦ê°€í•˜ë©° ê°’ì„ í•˜ë‚˜ì”© ë°˜í™˜í•˜ëŠ” ì œë„ˆë ˆì´í„° í•¨ìˆ˜
    :param start: ì‹œì‘ ê°’ (ê¸°ë³¸ê°’: 1)
    :param end: ë ê°’ (ê¸°ë³¸ê°’: 5)
    """
    i = start
    while i <= end:
        yield i  # ê°’ì„ í•˜ë‚˜ ë°˜í™˜í•˜ê³  í•¨ìˆ˜ ìƒíƒœë¥¼ ì €ì¥í•œ ì±„ ëŒ€ê¸°
        i += 1

gen = myrange(end=1000)  # myrange í•¨ìˆ˜ í˜¸ì¶œ â†’ ì œë„ˆë ˆì´í„° ê°ì²´ ìƒì„±

# next()ë¡œ ì œë„ˆë ˆì´í„°ì—ì„œ ê°’ í•˜ë‚˜ì”© êº¼ë‚´ê¸°
print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
print(next(gen))  # 4
print(next(gen))  # 5
````

````python
# forë¬¸ìœ¼ë¡œ ì œë„ˆë ˆì´í„° ìˆœíšŒ (1~9ê¹Œì§€ ì¶œë ¥)
for i in myrange(1, 10):
    print(i)
````

````python
def circle():
    """ì›ì˜ ë©´ì ì„ ê³„ì‚°í•˜ì—¬ ì¶œë ¥"""
    r = int(input("ë°˜ì§€ë¦„ : "))
    area = r * r * 3.14
    print(f"ë©´ì ì€ {area} ì…ë‹ˆë‹¤")

def rectangle():
    """ì‚¬ê°í˜•ì˜ ë©´ì ì„ ê³„ì‚°í•˜ì—¬ ì¶œë ¥"""
    width = int(input("ê°€ë¡œ : "))
    height = int(input("ì„¸ë¡œ : "))
    area = width * height
    print(f"ë©´ì ì€ {area} ì…ë‹ˆë‹¤")

def trapezoid():
    """ì‚¬ë‹¤ë¦¬ê¼´ì˜ ë©´ì ì„ ê³„ì‚°í•˜ì—¬ ì¶œë ¥"""
    base1 = int(input("ì•„ë«ë³€ : "))
    base2 = int(input("ìœ—ë³€ : "))
    height = int(input("ë†’ì´ : "))
    area = (base1 + base2) * height / 2
    print(f"ë©´ì ì€ {area} ì…ë‹ˆë‹¤")

def main_menu():
    """ë„í˜• ë©´ì  ê³„ì‚° ë©”ë‰´ í•¨ìˆ˜"""
    functions = {
        "1": circle,
        "2": rectangle,
        "3": trapezoid
    }
    while True:
        select = input("1.ì›ì˜ë©´ì  2.ì‚¬ê°í˜•ë©´ì  3.ì‚¬ë‹¤ë¦¬ê¼´ë©´ì  4.ì¢…ë£Œ ")
        if select in functions:
            functions[select]()  # ì„ íƒëœ í•¨ìˆ˜ í˜¸ì¶œ
        elif select == "4":
            return  # ì¢…ë£Œ
        else:
            print("ì˜¬ë°”ë¥¸ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")

# main_menu()
````

# ë¬¸ì œ2
ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ê°€ì„œ ë¦¬ìŠ¤íŠ¸ì•ˆì— ì¤‘ë³µëœ ë°ì´í„°ë¥¼ ì œê±°í•˜ê³  ì¤‘ë³µë˜ì§€ ì•ŠëŠ” ë°ì´í„° ë¦¬ìŠ¤íŠ¸ë§Œ ë°˜í™˜í•˜ê¸° 

````python
def duplicate_remove(a_list):
    """
    ë¦¬ìŠ¤íŠ¸ì—ì„œ ì¤‘ë³µëœ ê°’ì„ ì œê±°í•˜ê³ , ì¤‘ë³µë˜ì§€ ì•ŠëŠ” ê°’ë§Œ ë‚¨ê¸´ ìƒˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
    ì…ë ¥ ìˆœì„œë¥¼ ìœ ì§€í•©ë‹ˆë‹¤.
    :param a_list: ì¤‘ë³µì´ í¬í•¨ëœ ë¦¬ìŠ¤íŠ¸
    :return: ì¤‘ë³µì´ ì œê±°ëœ ë¦¬ìŠ¤íŠ¸
    """
    result = []
    for item in a_list:
        if item not in result:  # resultì— ì—†ëŠ” ê°’ë§Œ ì¶”ê°€
            result.append(item)
    return result

a = [4, 3, 5, 8, 1, 2, 56, 4, 1, 2, 8]
b = duplicate_remove(a)
print(b)
````

# ë¬¸ì œ 3
myint í•¨ìˆ˜ ë¬¸ìì—´ì„ ë°›ì•„ê°€ì„œ ì •ìˆ˜ë¡œ ë°”ê¾¸ì–´ì„œ ë°˜í™˜í•˜ê¸°. "123"ì„ ë„£ì—ˆì„ ê²½ìš°ì—ëŠ” 123ì„ 
ë°˜í™˜í•˜ê³  "123A" ì˜ëª»ëœ ë°ì´í„°ë¥¼ ì…ë ¥í•˜ë©´ -1ì„ ë°˜í™˜í•˜ì

````python
def myint(s):
    """
    ë¬¸ìì—´ sê°€ ì •ìˆ˜ë¡œë§Œ ì´ë£¨ì–´ì ¸ ìˆìœ¼ë©´ í•´ë‹¹ ì •ìˆ˜ê°’ì„ ë°˜í™˜,
    ìˆ«ìê°€ ì•„ë‹Œ ë¬¸ìê°€ í¬í•¨ë˜ì–´ ìˆìœ¼ë©´ -1ì„ ë°˜í™˜
    """
    result = 0
    for c in s:
        # ë¬¸ìê°€ 0~9 ì‚¬ì´ê°€ ì•„ë‹ˆë©´ -1 ë°˜í™˜
        if not ('0' <= c <= '9'):
            return -1
        # ì•„ìŠ¤í‚¤ ì½”ë“œë¡œ ë³€í™˜í•˜ì—¬ ìˆ«ìê°’ ëˆ„ì 
        result = result * 10 + (ord(c) - ord('0'))
    return result

print(myint("123") + myint("345"))
````

ë¬¸ì¥ì„ ë°›ì•„ê°€ì„œ ë¬¸ìì—´ ë’¤ì§‘ì–´ì„œ ë³´ë‚´ëŠ” í•¨ìˆ˜ reverse 

````python
def reverse(s):
    """
    ì…ë ¥ë°›ì€ ë¬¸ìì—´ së¥¼ ë’¤ì§‘ì–´ì„œ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
    :param s: ë¬¸ìì—´
    :return: ë’¤ì§‘íŒ ë¬¸ìì—´
    """
    result = ""
    # ë¬¸ìì—´ì˜ ë§ˆì§€ë§‰ ë¬¸ìë¶€í„° ì²« ë¬¸ìê¹Œì§€ ì—­ìˆœìœ¼ë¡œ resultì— ì¶”ê°€
    for i in range(len(s) - 1, -1, -1):
        result += s[i]
    return result

print(reverse("korea"))
````

````python
class Person:
    # í´ë˜ìŠ¤ ë³€ìˆ˜(ëª¨ë“  ì¸ìŠ¤í„´ìŠ¤ê°€ ê³µìœ , ì£¼ì˜: ê°€ë³€ ê°ì²´ëŠ” ì‚¬ìš© ì§€ì–‘)
    name = "í™ê¸¸ë™"
    age = 12
    phone = []  # í´ë˜ìŠ¤ ì „ì²´ì—ì„œ ê³µìœ ë¨ (ê°€ë³€ ê°ì²´ëŠ” ìƒì„±ìì—ì„œ ì„ ì–¸ ê¶Œì¥)

    def __init__(self):
        # ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜(ê° ê°ì²´ë§ˆë‹¤ ë³„ë„ ê³µê°„)
        self.name = ""
        self.age = 0
        self.phone = []  # ì¸ìŠ¤í„´ìŠ¤ë§ˆë‹¤ ë³„ë„ ë¦¬ìŠ¤íŠ¸

    def append(self, name="ì„êº½ì •", age=13, phone="010-0000-0001"):
        """
        ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ì— ê°’ í• ë‹¹ ë° ì „í™”ë²ˆí˜¸ ì¶”ê°€
        :param name: ì´ë¦„
        :param age: ë‚˜ì´
        :param phone: ì „í™”ë²ˆí˜¸(ë¬¸ìì—´)
        """
        self.name = name
        self.age = age
        self.phone.append(phone)  # ì¸ìŠ¤í„´ìŠ¤ë³„ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€

    def output(self):
        """ì¸ìŠ¤í„´ìŠ¤ì˜ ì •ë³´ë¥¼ ì¶œë ¥"""
        print(self.name, self.age, self.phone)

# ê°ì²´ ìƒì„± ë° ë°ì´í„° ì¶”ê°€
p1 = Person()
p1.append("ì¥ê¸¸ì‚°", 11, "010-0000-0003")

p2 = Person()
p2.append("ê¹€ì¢…ì„œ", 13, "010-0000-0004")

# ê²°ê³¼ ì¶œë ¥
p1.output()
p2.output()
````

````python
class Weekpay:
    """
    ì§ì›ì˜ ì£¼ê°„ ê¸‰ì—¬ë¥¼ ê³„ì‚°í•˜ëŠ” í´ë˜ìŠ¤
    """
    def __init__(self, name="", work_time=20, per_pay=10000):
        """
        :param name: ì§ì› ì´ë¦„ (ê¸°ë³¸ê°’: "")
        :param work_time: ê·¼ë¬´ ì‹œê°„ (ê¸°ë³¸ê°’: 20)
        :param per_pay: ì‹œê°„ë‹¹ ê¸‰ì—¬ (ê¸°ë³¸ê°’: 10000)
        """
        self.name = name
        self.work_time = work_time
        self.per_pay = per_pay
        self.process()  # ê¸‰ì—¬ ê³„ì‚° ë©”ì„œë“œ í˜¸ì¶œ

    def process(self):
        """
        ì£¼ê¸‰(pay) ê³„ì‚° (ê·¼ë¬´ ì‹œê°„ * ì‹œê°„ë‹¹ ê¸‰ì—¬)
        """
        self.pay = self.work_time * self.per_pay

    def output(self):
        """
        ì§ì›ì˜ ì •ë³´ë¥¼ ì¶œë ¥
        """
        print(f"{self.name} {self.work_time} {self.per_pay} {self.pay}")

# íŒŒì´ì¬ì˜ ë‚´ì¥ ë³€ìˆ˜ __name__ ì„¤ëª… ë° í…ŒìŠ¤íŠ¸ ì½”ë“œ
print(__name__)  # ì§ì ‘ ì‹¤í–‰ ì‹œ '__main__', import ì‹œ ëª¨ë“ˆëª…

if __name__ == "__main__":
    # í…ŒìŠ¤íŠ¸ìš© ê°ì²´ ìƒì„± ë° ì¶œë ¥
    w1 = Weekpay("A")
    w1.output()
````
