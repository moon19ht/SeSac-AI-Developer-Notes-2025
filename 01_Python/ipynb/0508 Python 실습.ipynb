````python
import random

class Baseball:
    """
    ìˆ«ì ì•¼êµ¬ ê²Œì„ í´ë˜ìŠ¤.
    - ì»´í“¨í„°ì™€ ì‚¬ìš©ìê°€ ê°ê° 3ìë¦¬ì˜ ì¤‘ë³µ ì—†ëŠ” ìˆ«ìë¥¼ ì„ íƒí•˜ì—¬ ìŠ¤íŠ¸ë¼ì´í¬/ë³¼/ì•„ì›ƒì„ íŒì •.
    - ìµœëŒ€ 5ë²ˆì˜ ê¸°íšŒ ë‚´ì— 3ìŠ¤íŠ¸ë¼ì´í¬ë¥¼ ë‹¬ì„±í•˜ë©´ ìŠ¹ë¦¬.
    """

    def __init__(self):
        # ì¸ë±ìŠ¤ 1~3ì— ì‹¤ì œ ê°’ ì €ì¥ (0ë²ˆ ì¸ë±ìŠ¤ëŠ” ì‚¬ìš©í•˜ì§€ ì•ŠìŒ)
        self.computer = [-1, -1, -1, -1]  # ì»´í“¨í„°ì˜ ìˆ«ì 3ê°œ
        self.person = [-1, -1, -1, -1]    # ì‚¬ìš©ìì˜ ìˆ«ì 3ê°œ
        self.count = 0                    # ì‹œë„ íšŸìˆ˜
        self.personList = []              # ê° ì‹œë„ ê²°ê³¼ ê¸°ë¡

    def init_computer(self):
        """
        ì»´í“¨í„°ì˜ 3ìë¦¬ ì¤‘ë³µ ì—†ëŠ” ìˆ«ì ìƒì„± (0~9).
        """
        cnt = 1
        while cnt <= 3:
            v = random.randint(0, 9)
            if v not in self.computer[1:cnt]:  # ì¤‘ë³µ ë°©ì§€
                self.computer[cnt] = v
                cnt += 1

    def init_person(self):
        """
        ì‚¬ìš©ìë¡œë¶€í„° 3ìë¦¬ ìˆ«ì ì…ë ¥ë°›ì•„ self.personì— ì €ì¥.
        """
        while True:
            s = input("ìˆ«ì 3ê°œ(0~9ì‚¬ì´ì˜)ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: 0 1 2): ")
            numberList = s.strip().split()
            if len(numberList) != 3:
                print("ì„¸ ê°œì˜ ìˆ«ìë¥¼ ê³µë°±ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ ì…ë ¥í•˜ì„¸ìš”.")
                continue
            try:
                nums = [int(x) for x in numberList]
            except ValueError:
                print("ìˆ«ìë§Œ ì…ë ¥í•˜ì„¸ìš”.")
                continue
            if len(set(nums)) != 3 or any(not (0 <= x <= 9) for x in nums):
                print("ì¤‘ë³µ ì—†ëŠ” 0~9 ì‚¬ì´ì˜ ìˆ«ì 3ê°œë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
                continue
            self.person[1], self.person[2], self.person[3] = nums
            break

    def get_result(self):
        """
        ìŠ¤íŠ¸ë¼ì´í¬, ë³¼, ì•„ì›ƒ ê°œìˆ˜ íŒì •.
        Returns:
            tuple: (strike, ball, out)
        """
        strike = 0
        ball = 0
        out = 0
        for i in range(1, 4):
            if self.person[i] == self.computer[i]:
                strike += 1
            elif self.person[i] in self.computer[1:]:
                ball += 1
            else:
                out += 1
        return strike, ball, out

    def start(self):
        """
        ê²Œì„ ì‹¤í–‰: 3ìŠ¤íŠ¸ë¼ì´í¬ ë˜ëŠ” 5íšŒ ì‹œë„ê¹Œì§€ ë°˜ë³µ.
        """
        self.init_computer()
        # print(self.computer)  # ë””ë²„ê¹…ìš©: ì •ë‹µ ê³µê°œ
        print("ì»´í“¨í„°ê°€ ìˆ«ìë¥¼ ì •í–ˆìŠµë‹ˆë‹¤. ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤!")
        while self.count < 5:
            self.init_person()
            strike, ball, out = self.get_result()
            print(f"{self.count+1}íšŒì°¨ ê²°ê³¼ â†’ Strike: {strike}, Ball: {ball}, Out: {out}")
            self.personList.append({
                "person": self.person[1:4].copy(),
                "strike": strike,
                "ball": ball,
                "out": out
            })
            self.count += 1
            if strike == 3:
                print(f"ì¶•í•˜í•©ë‹ˆë‹¤! {self.count}ë²ˆ ë§Œì— ë§ì·„ìŠµë‹ˆë‹¤!")
                break
        else:
            print(f"ì‹¤íŒ¨! ì •ë‹µì€ {self.computer[1]}, {self.computer[2]}, {self.computer[3]} ì˜€ìŠµë‹ˆë‹¤.")

if __name__ == "__main__":
    b = Baseball()
    b.start()
````

````python
class GameMain:
    """
    ìˆ«ì ì•¼êµ¬ ê²Œì„ì˜ ë©”ì¸ ì»¨íŠ¸ë¡¤ëŸ¬ í´ë˜ìŠ¤.
    - ê²Œì„ ì‹œì‘, í†µê³„ ì¶œë ¥, ì¢…ë£Œ ê¸°ëŠ¥ ì œê³µ
    - ì—¬ëŸ¬ ë²ˆì˜ ê²Œì„ ê²°ê³¼ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥
    """
    def __init__(self):
        self.gameList = []  # ê° ê²Œì„(Baseball ì¸ìŠ¤í„´ìŠ¤) ì €ì¥

    def start(self):
        """
        ë©”ì¸ ë©”ë‰´ ë£¨í”„: ê²Œì„ ì‹œì‘/í†µê³„/ì¢…ë£Œ ì„ íƒ
        """
        while True:
            print("\n===== ìˆ«ì ì•¼êµ¬ ê²Œì„ =====")
            print("1. ê²Œì„ ì‹œì‘")
            print("2. í†µê³„ ë³´ê¸°")
            print("0. ì¢…ë£Œ")
            sel = input("ì„ íƒ : ")
            if sel == "1":
                self.gamestart()
            elif sel == "2":
                self.showStatistics()
            elif sel == "0":
                print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
                return
            else:
                print("ì˜¬ë°”ë¥¸ ë²ˆí˜¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")

    def gamestart(self):
        """
        í•œ ë²ˆì˜ ìˆ«ì ì•¼êµ¬ ê²Œì„ì„ ì‹¤í–‰í•˜ê³  ê²°ê³¼ë¥¼ ì €ì¥
        """
        b = Baseball()      # Baseball ê²Œì„ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
        b.start()           # ê²Œì„ ì‹¤í–‰
        self.gameList.append(b)  # ê²°ê³¼ ì €ì¥

    def showStatistics(self):
        """
        ì§€ê¸ˆê¹Œì§€ í”Œë ˆì´í•œ ëª¨ë“  ê²Œì„ì˜ ê²°ê³¼(ì •ë‹µ, ì‹œë„ ë‚´ì—­) ì¶œë ¥
        """
        if not self.gameList:
            print("ì•„ì§ í”Œë ˆì´í•œ ê²Œì„ì´ ì—†ìŠµë‹ˆë‹¤.")
            return
        print("\n===== ê²Œì„ í†µê³„ =====")
        for idx, b in enumerate(self.gameList, 1):
            print(f"\n[{idx}ë²ˆì§¸ ê²Œì„] ì •ë‹µ: {b.computer[1]}, {b.computer[2]}, {b.computer[3]}")
            for i, item in enumerate(b.personList, 1):
                person_nums = ', '.join(map(str, item["person"]))
                print(f"  {i}íšŒì°¨: ì…ë ¥={person_nums} | "
                      f"Strike={item['strike']} Ball={item['ball']} Out={item['out']}")
            print(f"ì´ ì‹œë„ íšŸìˆ˜: {b.count}")

if __name__ == "__main__":
    g = GameMain()
    g.start()
````

````python
# í•œ ì‚¬ëŒì˜ ì„±ì  ì •ë³´ë¥¼ ì €ì¥í•˜ê³  ì²˜ë¦¬í•˜ëŠ” í´ë˜ìŠ¤
class ScoreData:
    def __init__(self, name="í™ê¸¸ë™", kor=100, eng=100, mat=100):
        """
        ìƒì„±ì: í•™ìƒ ì´ë¦„ê³¼ êµ­ì–´, ì˜ì–´, ìˆ˜í•™ ì ìˆ˜ë¥¼ ë°›ì•„ ì´ˆê¸°í™”
        Args:
            name (str): í•™ìƒ ì´ë¦„ (ê¸°ë³¸ê°’: "í™ê¸¸ë™")
            kor (int): êµ­ì–´ ì ìˆ˜ (ê¸°ë³¸ê°’: 100)
            eng (int): ì˜ì–´ ì ìˆ˜ (ê¸°ë³¸ê°’: 100)
            mat (int): ìˆ˜í•™ ì ìˆ˜ (ê¸°ë³¸ê°’: 100)
        """
        self.name = name
        self.kor = kor
        self.eng = eng
        self.mat = mat
        self.process()  # ì´ì , í‰ê· , í•™ì  ê³„ì‚°

    def process(self):
        """
        ì´ì , í‰ê· , í•™ì ì„ ê³„ì‚°í•˜ì—¬ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ë¡œ ì €ì¥
        """
        self.total = self.kor + self.eng + self.mat  # ì´ì  ê³„ì‚°
        self.avg = self.total / 3                    # í‰ê·  ê³„ì‚°
        # í•™ì  íŒì •
        if self.avg >= 90:
            self.grade = "ìˆ˜"
        elif self.avg >= 80:
            self.grade = "ìš°"
        elif self.avg >= 70:
            self.grade = "ë¯¸"
        elif self.avg >= 60:
            self.grade = "ì–‘"
        else:
            self.grade = "ê°€"

    def print(self):
        """
        í•™ìƒì˜ ì„±ì  ì •ë³´ë¥¼ íƒ­ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ ì¶œë ¥
        """
        print(f"{self.name}\t{self.kor}\t{self.eng}\t{self.mat}\t"
              f"{self.total}\t{self.avg:.2f}\t{self.grade}")

if __name__ == "__main__":
    # ê¸°ë³¸ê°’ìœ¼ë¡œ ê°ì²´ ìƒì„± ë° ì¶œë ¥ ì˜ˆì‹œ
    s = ScoreData()
    s.print()
````

````python
# from ScoreData import ScoreData
# ScoreData.py íŒŒì¼ì—ì„œ ScoreData í´ë˜ìŠ¤ë¥¼ ê°€ì ¸ì˜´

class ScoreManager:
    def __init__(self):
        """
        ìƒì„±ì: ScoreManager ê°ì²´ê°€ ìƒì„±ë  ë•Œ í˜¸ì¶œë¨.
        scoreListì— ì´ˆê¸° ScoreData ê°ì²´ë“¤ì„ ì €ì¥.
        """
        self.scoreList = [
            ScoreData(),
            ScoreData("ì¡°ìŠ¹ì—°", 90, 80, 90),
            ScoreData("ì•ˆì„¸ì˜", 80, 80, 70),
            ScoreData("ê¹€ì—°ê²½", 90, 90, 90),
            ScoreData("ê¹€ì—°ì•„", 100, 80, 100)
        ]

    def printAll(self):
        """
        scoreListì— ì €ì¥ëœ ëª¨ë“  ScoreData ê°ì²´ì˜ ì •ë³´ë¥¼ ì¶œë ¥.
        """
        for s in self.scoreList:
            s.print()

    def menuDisplay(self):
        """
        ì‚¬ìš©ìì—ê²Œ ë©”ë‰´ë¥¼ ì¶œë ¥.
        """
        print("--------")
        print("  ë©”ë‰´   ")
        print("--------")
        print("1. ì¶”ê°€  ")
        print("2. ì¶œë ¥  ")
        print("3. ê²€ìƒ‰  ")  # ì´ë¦„
        print("4. ìˆ˜ì •  ")  # ì´ë¦„
        print("5. ì‚­ì œ  ")  # ì´ë¦„
        print("6. ì •ë ¬  ")  # ì´ì  ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ
        print("0. ì¢…ë£Œ  ")
        print("--------")

    def append(self):
        """
        ìƒˆë¡œìš´ ScoreData ê°ì²´ë¥¼ ìƒì„±í•˜ì—¬ scoreListì— ì¶”ê°€.
        ì‚¬ìš©ìë¡œë¶€í„° ì´ë¦„, êµ­ì–´, ì˜ì–´, ìˆ˜í•™ ì ìˆ˜ë¥¼ ì…ë ¥ë°›ìŒ.
        """
        sc = ScoreData()
        sc.name = input("ì´ë¦„ : ")
        sc.kor = int(input("êµ­ì–´ : "))
        sc.eng = int(input("ì˜ì–´ : "))
        sc.mat = int(input("ìˆ˜í•™ : "))
        sc.process()  # ì´ì , í‰ê·  ë“± ê³„ì‚°
        self.scoreList.append(sc)

    def search(self):
        """
        ì´ë¦„ì„ ì…ë ¥ë°›ì•„ scoreListì—ì„œ í•´ë‹¹ ì´ë¦„ì´ í¬í•¨ëœ ScoreData ê°ì²´ë¥¼ ê²€ìƒ‰í•˜ì—¬ ì¶œë ¥.
        """
        name = input("ì´ë¦„ : ")
        # filterë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ë¦„ì´ í¬í•¨ëœ ê°ì²´ë§Œ ì¶”ì¶œ
        resultList = list(filter(lambda item: name in item.name, self.scoreList))
        if not resultList:
            print("ì°¾ìœ¼ì‹œëŠ” ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return

        for i, s in enumerate(resultList):
            print(f"[{i}] ", end=" ")
            s.print()

    def modify(self):
        """
        ì´ë¦„ì„ ì…ë ¥ë°›ì•„ í•´ë‹¹ ì´ë¦„ì´ í¬í•¨ëœ ScoreData ê°ì²´ë¥¼ ì°¾ì•„ ìˆ˜ì •.
        """
        name = input("ìˆ˜ì •í•  ì´ë¦„ : ")
        resultList = list(filter(lambda item: name in item.name, self.scoreList))
        if not resultList:
            print("ì°¾ìœ¼ì‹œëŠ” ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return

        for i, s in enumerate(resultList):
            print(f"[{i}] ", end=" ")
            s.print()
        sel = int(input("ìˆ˜ì •í•  ëŒ€ìƒì€(ë²ˆí˜¸) : "))
        # resultListì—ì„œ ì„ íƒí•œ ê°ì²´ì˜ ì°¸ì¡°ë¥¼ ê°€ì ¸ì˜´
        item = resultList[sel]
        item.name = input("ì´ë¦„ : ")
        item.kor = int(input("êµ­ì–´ : "))
        item.eng = int(input("ì˜ì–´ : "))
        item.mat = int(input("ìˆ˜í•™ : "))
        item.process()  # ì´ì , í‰ê·  ë“± ì¬ê³„ì‚°

    def delete(self):
        """
        ì´ë¦„ì„ ì…ë ¥ë°›ì•„ í•´ë‹¹ ì´ë¦„ì´ í¬í•¨ëœ ScoreData ê°ì²´ë¥¼ ì°¾ì•„ ì‚­ì œ.
        """
        name = input("ì‚­ì œí•  ì´ë¦„ : ")
        resultList = list(filter(lambda item: name in item.name, self.scoreList))
        if not resultList:
            print("ì°¾ìœ¼ì‹œëŠ” ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return

        for i, s in enumerate(resultList):
            print(f"[{i}] ", end=" ")
            s.print()
        sel = int(input("ì‚­ì œëŒ€ìƒì€(ë²ˆí˜¸) : "))
        # scoreListì—ì„œ í•´ë‹¹ ê°ì²´ë¥¼ ì‚­ì œ
        self.scoreList.remove(resultList[sel])

    def sort(self):
        """
        scoreListë¥¼ ì´ì  ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬í•˜ì—¬ ì¶œë ¥.
        ì›ë³¸ ë¦¬ìŠ¤íŠ¸ëŠ” ë³€ê²½í•˜ì§€ ì•ŠìŒ.
        """
        resultList = sorted(self.scoreList, key=lambda item: item.total, reverse=True)
        for s in resultList:
            s.print()

    def start(self):
        """
        í”„ë¡œê·¸ë¨ì˜ ë©”ì¸ ë£¨í”„.
        ë©”ë‰´ë¥¼ ì¶œë ¥í•˜ê³ , ì‚¬ìš©ìì˜ ì„ íƒì— ë”°ë¼ ê° ê¸°ëŠ¥ì„ ì‹¤í–‰.
        """
        funcList = [None, self.append, self.printAll,
                    self.search, self.modify,
                    self.delete, self.sort]
        while True:
            self.menuDisplay()
            try:
                choice = int(input("ì„ íƒ : "))
            except ValueError:
                print("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
                continue
            if 0 < choice < len(funcList):
                funcList[choice]()
            elif choice == 0:
                return
            else:
                print("ì˜ëª»ëœ ë©”ë‰´ì…ë‹ˆë‹¤.")

if __name__ == "__main__":
    sm = ScoreManager()
    sm.start()
````
