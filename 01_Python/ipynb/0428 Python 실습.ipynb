````python
# í•™ìƒ ì ìˆ˜ ê´€ë¦¬ í”„ë¡œê·¸ë¨ (í´ë¦° ì½”ë“œ: dict ë§¤í•‘ + 3ë‹¨ê³„ êµ¬ì¡°)

# ë“±ê¸‰ ì»·ì˜¤í”„ì™€ ë§¤í•‘
GRADE_CUT = {
    90: "ìˆ˜",
    80: "ìš°",
    70: "ë¯¸",
    60: "ì–‘",
     0: "ê°€"
}

# í•™ìƒ ìˆ˜ ì…ë ¥
student_count = int(input("í•™ìƒ ìˆ˜ : "))
students = []

# [1] ì…ë ¥ ë‹¨ê³„: ì´ë¦„ê³¼ ì ìˆ˜ë§Œ ì €ì¥
for _ in range(student_count):
    name = input("ì´ë¦„   : ")
    while True:
        try:
            kor = int(input("êµ­ì–´   : "))
            break
        except ValueError:
            print("ì •ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
    while True:
        try:
            eng = int(input("ì˜ì–´   : "))
            break
        except ValueError:
            print("ì •ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
    while True:
        try:
            math = int(input("ìˆ˜í•™   : "))
            break
        except ValueError:
            print("ì •ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
    students.append({
        "name": name,
        "kor": kor,
        "eng": eng,
        "math": math
    })

# [2] ì²˜ë¦¬ ë‹¨ê³„: ì´ì Â·í‰ê·  ê³„ì‚°, dict ë§¤í•‘ìœ¼ë¡œ ë“±ê¸‰ ê²°ì •
for s in students:
    s["total"] = s["kor"] + s["eng"] + s["math"]
    s["avg"]   = s["total"] / 3

    # dict ìˆœíšŒí•˜ë©´ì„œ ì²« ë²ˆì§¸ ë§¤ì¹­ëœ ë“±ê¸‰ìœ¼ë¡œ ê²°ì •
    for cutoff, grade in GRADE_CUT.items():
        if s["avg"] >= cutoff:
            s["grade"] = grade
            break

# [3] ì¶œë ¥ ë‹¨ê³„
print("ì´ë¦„\têµ­ì–´\tì˜ì–´\tìˆ˜í•™\tì´ì \tí‰ê· \të“±ê¸‰")
for s in students:
    print(f"{s['name']}\t{s['kor']}\t{s['eng']}\t{s['math']}\t{s['total']}\t{s['avg']:.2f}\t{s['grade']}")
````
