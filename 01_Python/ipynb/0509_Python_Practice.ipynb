# mpg.csv ì½ê¸°

````python
# mpg.csv íŒŒì¼ì—ì„œ ê° ì‹¤ë¦°ë”(cylinders)ë³„ ê°œìˆ˜ë¥¼ ì„¸ëŠ” ì½”ë“œì…ë‹ˆë‹¤.
with open("../data/mpg.csv", "r") as f:
    lines = f.readlines()

# ì²« ì¤„(í—¤ë”)ì„ ì œì™¸í•œ ë°ì´í„°ë§Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
data_lines = lines[1:]  # í—¤ë”ë¥¼ ì œì™¸í•œ ë°ì´í„° ë¶€ë¶„

cylinder_count = {}  # ì‹¤ë¦°ë”ë³„ ê°œìˆ˜ë¥¼ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬

for line in data_lines:
    line = line.rstrip('\n')  # ì¤„ ëì˜ ê°œí–‰ ë¬¸ì ì œê±°
    values = line.split(",")  # ì½¤ë§ˆë¡œ ë¶„ë¦¬í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
    cyl = values[1]           # ë‘ ë²ˆì§¸ ê°’ì´ ì‹¤ë¦°ë” ìˆ˜
    if cyl in cylinder_count:
        cylinder_count[cyl] += 1
    else:
        cylinder_count[cyl] = 1

print(cylinder_count)
````

# iris.csv ì½ê¸°

````python
# iris.csv íŒŒì¼ì„ ì½ì–´ì„œ 4ê°œ ì»¬ëŸ¼ì˜ í‰ê· ì„ êµ¬í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤.
# í•œê¸€ì²˜ë¦¬: cp949(ìœˆë„ìš°) ë˜ëŠ” utf-8(í‘œì¤€, vscodeëŠ” utf-8)
irisList = []  # ê° í–‰ì˜ 4ê°œ ìˆ˜ì¹˜ ë°ì´í„°ê°€ ë“¤ì–´ê°ˆ ë¦¬ìŠ¤íŠ¸

with open("../data/iris.csv", "r", encoding="utf-8") as f:
    lines = f.readlines()

# ì²« ì¤„ì€ í—¤ë”ì´ë¯€ë¡œ 1ë²ˆ ì¸ë±ìŠ¤ë¶€í„° ë°˜ë³µ
for i in range(1, len(lines)):
    line = lines[i].rstrip('\n')  # ê°œí–‰ ë¬¸ì ì œê±°
    print(i, line)
    values = line.split(",")
    # ì•ì˜ 4ê°œ ì»¬ëŸ¼ë§Œ float ë³€í™˜í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥
    data = [float(values[0]), float(values[1]), float(values[2]), float(values[3])]
    irisList.append(data)

# ê° í–‰ ì¶œë ¥
for iris in irisList:
    print(iris)
# print(irisList)  # ì „ì²´ ë¦¬ìŠ¤íŠ¸ ì¶œë ¥(ì£¼ì„ ì²˜ë¦¬)

# 4ê°œ ì»¬ëŸ¼ë³„ í•©ê³„ êµ¬í•˜ê¸°
result = [0, 0, 0, 0]
for j in range(4):
    for i in range(len(irisList)):
        result[j] += irisList[i][j]

# ì „ì²´ í‰ê·  ì¶œë ¥ (ì´ 150ê°œ ë°ì´í„°)
print(result[0]/150, result[1]/150, result[2]/150, result[3]/150)

# ì†Œìˆ˜ì  2ìë¦¬ë¡œ í‰ê·  ì¶œë ¥ (íƒ­ êµ¬ë¶„)
count = len(irisList)
for i in range(4):
    print(f"{result[i]/count:.2f}", end="\t")
print()

# mpg.csvíŒŒì¼ ê°€ì ¸ì™€ì„œ ì‹¤ë¦°ë”ê°œìˆ˜ 8 6 4 ì¢…ë¥˜ë³„ ì¹´ìš´íŠ¸ í•˜ê¸°
````

# ì •ìˆ˜ íŒŒì¼ ì½ê¸°

````python
# ì •ìˆ˜ íŒŒì¼ì„ ì½ì–´ì„œ í•©ê³„ì™€ í‰ê· ì„ êµ¬í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤.
with open("../data/integer.txt", "r") as f:
    lines = f.readlines()  # íŒŒì¼ì˜ ëª¨ë“  ì¤„ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ì½ìŒ

total = 0  # í•©ê³„ ì €ì¥ ë³€ìˆ˜
for line in lines:
    line = line.strip()  # ê°œí–‰ ë¬¸ì ë“± ì–‘ìª½ ê³µë°± ì œê±°
    print(line)          # ê° ì¤„(ì •ìˆ˜) ì¶œë ¥
    total += int(line)   # ì •ìˆ˜ë¡œ ë³€í™˜í•˜ì—¬ í•©ê³„ì— ë”í•¨

# í‰ê·  ê³„ì‚° ë° ì¶œë ¥
print("í‰ê· ", total / len(lines))
````

# ì„±ì  íŒŒì¼ ì½ê¸°

````python
# score.txt íŒŒì¼ì„ ì½ì–´ì„œ ê° í•™ìƒì˜ ì´ë¦„, ì ìˆ˜, ì´ì , í‰ê· ì„ ì¶œë ¥í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤.
# í•œê¸€ ì¸ì½”ë”©: cp949(ìœˆë„ìš°) ë˜ëŠ” utf-8(í‘œì¤€, vscodeëŠ” utf-8)
with open("../data/score.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()  # íŒŒì¼ì˜ ëª¨ë“  ì¤„ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ì½ìŒ

for line in lines:
    line = line.strip()  # ì–‘ìª½ ê³µë°± ë° ê°œí–‰ ë¬¸ì ì œê±°
    words = line.split(",")  # ì½¤ë§ˆë¡œ ë¶„ë¦¬í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
    name = words[0]
    kor = int(words[1])
    eng = int(words[2])
    mat = int(words[3])
    total = kor + eng + mat  # ì´ì  ê³„ì‚°
    avg = total / 3          # í‰ê·  ê³„ì‚°
    print(name, kor, eng, mat, total, avg)  # ê²°ê³¼ ì¶œë ¥
````
