- **ë¶€ëª¨ í´ë˜ìŠ¤(Parent)** ì˜ ì†ì„±ê³¼ ë©”ì„œë“œë¥¼ **ìì‹ í´ë˜ìŠ¤(Child)** ê°€ ë¬¼ë ¤ë°›ìŒ

## âœ… ë¬¸ë²• ë° ì˜ˆì‹œ
```python
class Animal:
    def speak(self):
        print("ë™ë¬¼ì´ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤.")
```

````python
class Base:  # ë¶€ëª¨ í´ë˜ìŠ¤
    def __init__(self, x=0, y=0):
        # x, y ì¢Œí‘œê°’ì„ ì´ˆê¸°í™”
        self.x = x
        self.y = y
        print("Base ìƒì„±ì")

    def display(self):
        # í˜„ì¬ ê°ì²´ì˜ x, y ê°’ì„ ì¶œë ¥
        print(f'x={self.x} y={self.y}')

    def add(self):
        # xì™€ yì˜ í•©ì„ ë°˜í™˜
        return self.x + self.y

    def doubleX(self):
        # xì˜ ë‘ ë°° ê°’ì„ ë°˜í™˜
        return self.x * 2

    def doubleY(self):
        # yì˜ ë‘ ë°° ê°’ì„ ë°˜í™˜
        return self.y * 2
````

# ë‹¤í˜•ì„±(Polymorphism)
- ì˜¤ë²„ë¡œë”©(Overloading): ê°™ì€ ì´ë¦„ì˜ ë©”ì„œë“œë¥¼ ë§¤ê°œë³€ìˆ˜ í˜•íƒœë§Œ ë‹¤ë¥´ê²Œ ì—¬ëŸ¬ ê°œ ì •ì˜ (íŒŒì´ì¬ì€ ì§€ì›í•˜ì§€ ì•ŠìŒ)
- ì˜¤ë²„ë¼ì´ë”©(Overriding): ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì„œë“œë¥¼ ìì‹ í´ë˜ìŠ¤ì—ì„œ ì¬ì •ì˜

````python
class Child1(Base):  # Base í´ë˜ìŠ¤ë¥¼ ìƒì†ë°›ëŠ” ìì‹ í´ë˜ìŠ¤
    def __init__(self, x=0, y=0, z=0):
        # ë¶€ëª¨ ìƒì„±ìë¥¼ í˜¸ì¶œí•˜ì—¬ x, y ì´ˆê¸°í™”
        super().__init__(x, y)
        # ìì‹ í´ë˜ìŠ¤ì—ì„œë§Œ ì‚¬ìš©í•˜ëŠ” z ë³€ìˆ˜ ì¶”ê°€
        self.z = z
        print("Child1 ìƒì„±ì")

    def display(self):
        # ë¶€ëª¨ì˜ displayë¥¼ í˜¸ì¶œí•˜ì§€ ì•Šê³ , x, y, z ëª¨ë‘ ì¶œë ¥
        print(f"x={self.x} y={self.y} z={self.z}")
````

````python
# Base í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ë° ë©”ì„œë“œ í˜¸ì¶œ ì˜ˆì‹œ
p = Base()  # x=0, y=0
p.display()

p = Base(4, 5)  # x=4, y=5
p.display()

# Child1 í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ë° ë©”ì„œë“œ í˜¸ì¶œ ì˜ˆì‹œ
c1 = Child1(1, 2, 3)  # x=1, y=2, z=3
c1.display()
print(c1.doubleX())  # xì˜ ë‘ ë°° ì¶œë ¥
print(c1.doubleY())  # yì˜ ë‘ ë°° ì¶œë ¥
````

# ğŸ“Œ 2. ë‹¤ì¤‘ ìƒì† (Multiple Inheritance)

## âœ… ê°œë…
- ë‘ ê°œ ì´ìƒì˜ í´ë˜ìŠ¤ë¡œë¶€í„° **ì†ì„±ê³¼ ë©”ì„œë“œ**ë¥¼ ë™ì‹œì— ìƒì†ë°›ì„ ìˆ˜ ìˆìŒ
- ìˆœì„œì— ë”°ë¼ **MRO(Method Resolution Order)** ì ìš©

## âœ… ì˜ˆì‹œ
```python
class A:
    def do_a(self):
        print("A ë™ì‘")

class B:
    def do_b(self):
        print("B ë™ì‘")

class C(A, B):
    pass

c = C()
c.do_a()
c.do_b()
```

> `C`ëŠ” `A`, `B` ëª¨ë‘ë¥¼ ìƒì†ë°›ì•„ ë‘ í´ë˜ìŠ¤ì˜ ê¸°ëŠ¥ì„ ë™ì‹œì— ì‚¬ìš©í•  ìˆ˜ ìˆìŒ.

````python
class Flyable:
    def fly(self):
        print("ë‚  ìˆ˜ ìˆë‹¤")  # Can fly

    def walk(self):
        print("ë‘ë‹¤ë¦¬ë¡œ ê±·ëŠ”ë‹¤")  # Walks on two legs (Flyable version)


class Swimmable:
    def swim(self):
        print("ìˆ˜ì˜í•  ìˆ˜ ìˆë‹¤")  # Can swim

    def walk(self):
        print("*** ë‘ë‹¤ë¦¬ë¡œ ê±·ëŠ”ë‹¤ ***")  # Walks on two legs (Swimmable version)


class Duck(Swimmable, Flyable):  # ë‹¤ì¤‘ìƒì†: Swimmable, Flyable ìˆœì„œ
    def quack(self):
        print("ê½¥ê½¥")  # Quack


d1 = Duck()  # Duck ì¸ìŠ¤í„´ìŠ¤ ìƒì„±

d1.fly()    # Flyableì˜ fly() í˜¸ì¶œ
d1.swim()   # Swimmableì˜ swim() í˜¸ì¶œ
d1.quack()  # Duckì˜ quack() í˜¸ì¶œ

# walk() ë©”ì„œë“œëŠ” Swimmableì´ ë¨¼ì € ìƒì†ë˜ì—ˆìœ¼ë¯€ë¡œ Swimmableì˜ walk()ê°€ í˜¸ì¶œë¨
d1.walk()

# __mro__ëŠ” í´ë˜ìŠ¤ì˜ ìƒì† ê´€ê³„(ë©”ì„œë“œ íƒìƒ‰ ìˆœì„œ)ë¥¼ íŠœí”Œë¡œ ë°˜í™˜
print(Duck.__mro__)
````

# ğŸ“Œ 3. ë‹¤ì´ì•„ëª¬ë“œ ìƒì† (Diamond Inheritance)

## âœ… ê°œë…
- ìƒì† êµ¬ì¡°ê°€ **ë§ˆë¦„ëª¨ í˜•íƒœ**ë¡œ êµ¬ì„±ë˜ì–´, **ê°™ì€ ì¡°ìƒ í´ë˜ìŠ¤ê°€ ì¤‘ë³µ ìƒì†ë˜ëŠ” êµ¬ì¡°**
- Pythonì€ **MRO(Method Resolution Order)** ê·œì¹™ì— ë”°ë¼ ì¤‘ë³µì„ í•´ê²°

## âœ… ì˜ˆì‹œ
```python
class A:
    def whoami(self):
        print("A")

class B(A):
    def whoami(self):
        print("B")

class C(A):
    def whoami(self):
        print("C")

class D(B, C):
    pass

d = D()
d.whoami()  # ì¶œë ¥: B â†’ MROì— ë”°ë¼ ì¢Œì¸¡ ìš°ì„ 
```

> `D -> B -> C -> A` ìˆœìœ¼ë¡œ ë©”ì„œë“œ íƒìƒ‰ (ì¢Œì¸¡ í´ë˜ìŠ¤ ìš°ì„ )

````python
class A:
    def __init__(self):
        print("A ìƒì„±ì í˜¸ì¶œ")  # Aì˜ ìƒì„±ì

class B(A):  # Aë¥¼ ìƒì†ë°›ìŒ
    def __init__(self):
        print("B ìƒì„±ì í˜¸ì¶œ")  # Bì˜ ìƒì„±ì
        super().__init__()     # ë¶€ëª¨(A) ìƒì„±ì í˜¸ì¶œ

class C(A):  # Aë¥¼ ìƒì†ë°›ìŒ
    def __init__(self):
        print("C ìƒì„±ì í˜¸ì¶œ")  # Cì˜ ìƒì„±ì
        super().__init__()     # ë¶€ëª¨(A) ìƒì„±ì í˜¸ì¶œ

class D(B, C):  # Bì™€ Cë¥¼ ëª¨ë‘ ìƒì†ë°›ìŒ (ë‹¤ì´ì•„ëª¬ë“œ ìƒì† êµ¬ì¡°)
    def __init__(self):
        print("D ìƒì„±ì í˜¸ì¶œ")  # Dì˜ ìƒì„±ì
        super().__init__()     # ë¶€ëª¨(B, C) ìƒì„±ì í˜¸ì¶œ (MROì— ë”°ë¼ ìˆœì„œ ê²°ì •)
````

````python
d = D()  # D ì¸ìŠ¤í„´ìŠ¤ ìƒì„±, ìƒì„±ì í˜¸ì¶œ ìˆœì„œëŠ” MRO(Method Resolution Order)ì— ë”°ë¦„

# isinstance: ê°ì²´ê°€ í•´ë‹¹ í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ì¸ì§€ í™•ì¸
print(isinstance(d, A))      # dëŠ” Aì˜ ì¸ìŠ¤í„´ìŠ¤ì¸ê°€?
print(isinstance(d, B))      # dëŠ” Bì˜ ì¸ìŠ¤í„´ìŠ¤ì¸ê°€?
print(isinstance(d, C))      # dëŠ” Cì˜ ì¸ìŠ¤í„´ìŠ¤ì¸ê°€?
print(isinstance(d, object)) # dëŠ” objectì˜ ì¸ìŠ¤í„´ìŠ¤ì¸ê°€? (ëª¨ë“  í´ë˜ìŠ¤ì˜ ìµœìƒìœ„)
print(isinstance(d, str))    # dëŠ” strì˜ ì¸ìŠ¤í„´ìŠ¤ì¸ê°€?

# object: ëª¨ë“  í´ë˜ìŠ¤ì˜ base í´ë˜ìŠ¤
a = object()
print(a.__class__)  # objectì˜ í´ë˜ìŠ¤ ì •ë³´ ì¶œë ¥
````
