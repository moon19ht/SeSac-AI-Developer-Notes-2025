## 1.1 ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œ

### ê°œë…
- **ë”•ì…”ë„ˆë¦¬(Dictionary)**ë¥¼ í™œìš©í•œ í•™ìƒ ì •ë³´ ì €ì¥
- **ë¦¬ìŠ¤íŠ¸(List)**ë¥¼ í™œìš©í•œ ì—¬ëŸ¬ í•™ìƒ ê´€ë¦¬
- ë“±ê¸‰ ê³„ì‚° ë¡œì§ êµ¬í˜„

### ë“±ê¸‰ ê¸°ì¤€
- ìˆ˜(90ì  ì´ìƒ), ìš°(80ì  ì´ìƒ), ë¯¸(70ì  ì´ìƒ), ì–‘(60ì  ì´ìƒ), ê°€(60ì  ë¯¸ë§Œ)

```python
# ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œ - ê¸°ë³¸ ë²„ì „
# ëª©í‘œ: ë”•ì…”ë„ˆë¦¬ì™€ ë¦¬ìŠ¤íŠ¸ë¥¼ í™œìš©í•œ ê¸°ë³¸ì ì¸ ì„±ì  ê´€ë¦¬

def grade_management_basic():
    students = []
    
    # í•™ìƒ ìˆ˜ ì…ë ¥
    student_count = int(input("í•™ìƒ ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
    
    # í•™ìƒ ì •ë³´ ì…ë ¥
    for i in range(student_count):
        print(f"\n{i+1}ë²ˆì§¸ í•™ìƒ ì •ë³´ ì…ë ¥")
        name = input("ì´ë¦„: ")
        korean = int(input("êµ­ì–´: "))
        english = int(input("ì˜ì–´: "))
        math = int(input("ìˆ˜í•™: "))
        
        student = {
            "name": name,
            "korean": korean,
            "english": english,
            "math": math
        }
        students.append(student)
    
    # ì´ì , í‰ê· , ë“±ê¸‰ ê³„ì‚°
    for student in students:
        total = student["korean"] + student["english"] + student["math"]
        avg = total / 3
        
        # ë“±ê¸‰ íŒì •
        if avg >= 90:
            grade = "ìˆ˜"
        elif avg >= 80:
            grade = "ìš°"
        elif avg >= 70:
            grade = "ë¯¸"
        elif avg >= 60:
            grade = "ì–‘"
        else:
            grade = "ê°€"
        
        # ê²°ê³¼ ì¶”ê°€
        student["total"] = total
        student["avg"] = avg
        student["grade"] = grade
    
    # ê²°ê³¼ ì¶œë ¥
    print("\n" + "="*60)
    print("ì„±ì í‘œ")
    print("="*60)
    print(f"{'ì´ë¦„':^8} {'êµ­ì–´':^6} {'ì˜ì–´':^6} {'ìˆ˜í•™':^6} {'ì´ì ':^6} {'í‰ê· ':^8} {'ë“±ê¸‰':^4}")
    print("-"*60)
    
    for student in students:
        print(f"{student['name']:^8} {student['korean']:^6} {student['english']:^6} "
              f"{student['math']:^6} {student['total']:^6} {student['avg']:^8.1f} {student['grade']:^4}")

# ì‹¤í–‰
# grade_management_basic()
```

```python
# ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œ - ê°œì„  ë²„ì „ (í´ë¦° ì½”ë“œ)
# ëª©í‘œ: ë”•ì…”ë„ˆë¦¬ ë§¤í•‘ì„ í™œìš©í•œ ê¹”ë”í•œ ì½”ë“œ, ì˜ˆì™¸ ì²˜ë¦¬ í¬í•¨

def grade_management_advanced():
    # ë“±ê¸‰ ì»·ì˜¤í”„ ë§¤í•‘
    GRADE_CUTOFF = {
        90: "ìˆ˜",
        80: "ìš°",
        70: "ë¯¸",
        60: "ì–‘",
        0: "ê°€"
    }
    
    def get_grade(avg):
        """í‰ê·  ì ìˆ˜ì— ë”°ë¥¸ ë“±ê¸‰ ë°˜í™˜"""
        for cutoff, grade in GRADE_CUTOFF.items():
            if avg >= cutoff:
                return grade
        return "ê°€"
    
    def input_score(subject):
        """ê³¼ëª©ë³„ ì ìˆ˜ ì…ë ¥ (ì˜ˆì™¸ ì²˜ë¦¬ í¬í•¨)"""
        while True:
            try:
                score = int(input(f"{subject}: "))
                if 0 <= score <= 100:
                    return score
                else:
                    print("0~100 ì‚¬ì´ì˜ ì ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
            except ValueError:
                print("ì •ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”.")
    
    students = []
    
    # [1] ì…ë ¥ ë‹¨ê³„
    student_count = int(input("í•™ìƒ ìˆ˜: "))
    
    for i in range(student_count):
        print(f"\n{i+1}ë²ˆì§¸ í•™ìƒ ì •ë³´")
        name = input("ì´ë¦„: ")
        korean = input_score("êµ­ì–´")
        english = input_score("ì˜ì–´")
        math = input_score("ìˆ˜í•™")
        
        students.append({
            "name": name,
            "korean": korean,
            "english": english,
            "math": math
        })
    
    # [2] ì²˜ë¦¬ ë‹¨ê³„
    for student in students:
        student["total"] = student["korean"] + student["english"] + student["math"]
        student["avg"] = student["total"] / 3
        student["grade"] = get_grade(student["avg"])
    
    # [3] ì¶œë ¥ ë‹¨ê³„
    print("\n" + "="*65)
    print("ğŸ“Š ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œ ê²°ê³¼")
    print("="*65)
    print(f"{'ì´ë¦„':^8} {'êµ­ì–´':^6} {'ì˜ì–´':^6} {'ìˆ˜í•™':^6} {'ì´ì ':^6} {'í‰ê· ':^8} {'ë“±ê¸‰':^4}")
    print("-"*65)
    
    for student in students:
        print(f"{student['name']:^8} {student['korean']:^6} {student['english']:^6} "
              f"{student['math']:^6} {student['total']:^6} {student['avg']:^8.2f} {student['grade']:^4}")
    
    # í†µê³„ ì •ë³´
    if students:
        total_avg = sum(s["avg"] for s in students) / len(students)
        print(f"\nğŸ“ˆ ì „ì²´ í‰ê· : {total_avg:.2f}ì ")
        
        best_student = max(students, key=lambda x: x["avg"])
        print(f"ğŸ† ìµœê³ ì : {best_student['name']} ({best_student['avg']:.2f}ì )")

# ì‹¤í–‰
# grade_management_advanced()
```

## 1.2 ê¸‰ì—¬ ê´€ë¦¬ ì‹œìŠ¤í…œ

### ê°œë…
- ì§ì›ì˜ ê·¼ë¬´ì‹œê°„ê³¼ ì‹œê°„ë‹¹ ê¸‰ì—¬ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì£¼ê¸‰ ê³„ì‚°
- ë¦¬ìŠ¤íŠ¸ì™€ ë”•ì…”ë„ˆë¦¬ë¥¼ í™œìš©í•œ ë°ì´í„° ê´€ë¦¬

```python
# ê¸‰ì—¬ ê´€ë¦¬ ì‹œìŠ¤í…œ - ê¸°ë³¸ ë²„ì „ (ë¦¬ìŠ¤íŠ¸ í™œìš©)
def salary_management_basic():
    print("ğŸ“‹ ê¸‰ì—¬ ê´€ë¦¬ ì‹œìŠ¤í…œ (ë¦¬ìŠ¤íŠ¸ ë²„ì „)")
    
    employee_count = int(input("ì§ì› ìˆ˜: "))
    
    names = []
    work_times = []
    hourly_pays = []
    
    # ì…ë ¥
    for i in range(employee_count):
        print(f"\n{i+1}ë²ˆì§¸ ì§ì› ì •ë³´")
        name = input("ì´ë¦„: ")
        work_time = int(input("ê·¼ë¬´ì‹œê°„: "))
        hourly_pay = int(input("ì‹œê°„ë‹¹ ê¸‰ì—¬: "))
        
        names.append(name)
        work_times.append(work_time)
        hourly_pays.append(hourly_pay)
    
    # ì¶œë ¥
    print("\n" + "="*50)
    print("ğŸ’° ê¸‰ì—¬ ê³„ì‚° ê²°ê³¼")
    print("="*50)
    print(f"{'ì´ë¦„':^10} {'ê·¼ë¬´ì‹œê°„':^8} {'ì‹œê¸‰':^10} {'ì£¼ê¸‰':^12}")
    print("-"*50)
    
    total_salary = 0
    for i in range(employee_count):
        weekly_pay = work_times[i] * hourly_pays[i]
        total_salary += weekly_pay
        print(f"{names[i]:^10} {work_times[i]:^8} {hourly_pays[i]:^10,} {weekly_pay:^12,}")
    
    print(f"\nğŸ’µ ì´ ì§€ê¸‰ì•¡: {total_salary:,}ì›")

# ì‹¤í–‰
# salary_management_basic()
```

```python
# ê¸‰ì—¬ ê´€ë¦¬ ì‹œìŠ¤í…œ - ê°œì„  ë²„ì „ (ë”•ì…”ë„ˆë¦¬ í™œìš©)
def salary_management_advanced():
    print("ğŸ“‹ ê¸‰ì—¬ ê´€ë¦¬ ì‹œìŠ¤í…œ (ë”•ì…”ë„ˆë¦¬ ë²„ì „)")
    
    # ê¸°ë³¸ ì§ì› ë°ì´í„° (ì˜ˆì‹œ)
    employees = [
        {"name": "í™ê¸¸ë™", "work_time": 40, "hourly_pay": 12000},
        {"name": "ì„êº½ì •", "work_time": 35, "hourly_pay": 15000},
        {"name": "ì¥ê¸¸ì‚°", "work_time": 30, "hourly_pay": 18000},
    ]
    
    # ì¶”ê°€ ì§ì› ì…ë ¥
    additional = int(input("ì¶”ê°€ë¡œ ì…ë ¥í•  ì§ì› ìˆ˜: "))
    
    for i in range(additional):
        print(f"\n{i+1}ë²ˆì§¸ ì¶”ê°€ ì§ì›")
        name = input("ì´ë¦„: ")
        work_time = int(input("ê·¼ë¬´ì‹œê°„: "))
        hourly_pay = int(input("ì‹œê°„ë‹¹ ê¸‰ì—¬: "))
        
        employees.append({
            "name": name,
            "work_time": work_time,
            "hourly_pay": hourly_pay
        })
    
    # ì£¼ê¸‰ ê³„ì‚°
    for employee in employees:
        employee["weekly_pay"] = employee["work_time"] * employee["hourly_pay"]
        
        # ì´ˆê³¼ ê·¼ë¬´ ìˆ˜ë‹¹ (40ì‹œê°„ ì´ˆê³¼ì‹œ 1.5ë°°)
        if employee["work_time"] > 40:
            overtime = employee["work_time"] - 40
            employee["overtime_pay"] = overtime * employee["hourly_pay"] * 0.5
            employee["total_pay"] = employee["weekly_pay"] + employee["overtime_pay"]
        else:
            employee["overtime_pay"] = 0
            employee["total_pay"] = employee["weekly_pay"]
    
    # ê²°ê³¼ ì¶œë ¥
    print("\n" + "="*80)
    print("ğŸ’° ê¸‰ì—¬ ê³„ì‚° ê²°ê³¼ (ì´ˆê³¼ê·¼ë¬´ ìˆ˜ë‹¹ í¬í•¨)")
    print("="*80)
    print(f"{'ì´ë¦„':^8} {'ê·¼ë¬´ì‹œê°„':^8} {'ì‹œê¸‰':^10} {'ê¸°ë³¸ê¸‰':^12} {'ì´ˆê³¼ìˆ˜ë‹¹':^12} {'ì´ ê¸‰ì—¬':^12}")
    print("-"*80)
    
    total_cost = 0
    for emp in employees:
        total_cost += emp["total_pay"]
        print(f"{emp['name']:^8} {emp['work_time']:^8} {emp['hourly_pay']:^10,} "
              f"{emp['weekly_pay']:^12,} {emp['overtime_pay']:^12,} {emp['total_pay']:^12,}")
    
    print(f"\nğŸ’µ ì´ ì¸ê±´ë¹„: {total_cost:,}ì›")
    print(f"ğŸ“Š í‰ê·  ê¸‰ì—¬: {total_cost/len(employees):,.0f}ì›")

# ì‹¤í–‰
# salary_management_advanced()
```

## 1.3 ì„±ë³„ ì„ê¸ˆ ë¶„ì„ ì‹œìŠ¤í…œ

### ê°œë…
- ì„±ë³„ì— ë”°ë¥¸ ì„ê¸ˆ ì°¨ì´ ë¶„ì„
- ê·¸ë£¹ë³„ í†µê³„ ê³„ì‚°

```python
# ì„±ë³„ ì„ê¸ˆ ë¶„ì„ ì‹œìŠ¤í…œ
def gender_salary_analysis():
    print("ğŸ“Š ì„±ë³„ ì„ê¸ˆ ë¶„ì„ ì‹œìŠ¤í…œ")
    print("ë…¸ë™ì²­ ì‹ ê³  ëŒ€ì‘ì„ ìœ„í•œ ì„±ë³„ ì„ê¸ˆ í‰ê·  ë¶„ì„")
    
    employee_count = int(input("\në¶„ì„í•  ì§ì› ìˆ˜: "))
    employees = []
    
    # ì§ì› ì •ë³´ ì…ë ¥
    for i in range(employee_count):
        print(f"\n{i+1}ë²ˆì§¸ ì§ì› ì •ë³´")
        
        # ì„±ë³„ ì…ë ¥ (ìœ íš¨ì„± ê²€ì‚¬)
        while True:
            gender = input("ì„±ë³„ (ë‚¨/ì—¬): ").strip()
            if gender in ['ë‚¨', 'ì—¬']:
                break
            print("'ë‚¨' ë˜ëŠ” 'ì—¬'ë¡œ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        
        # ì—°ë´‰ ì…ë ¥ (ìœ íš¨ì„± ê²€ì‚¬)
        while True:
            try:
                salary = int(input("ì—°ë´‰ (ë§Œì›): "))
                if salary > 0:
                    break
                print("ì–‘ìˆ˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
            except ValueError:
                print("ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        
        employees.append({
            "gender": gender,
            "salary": salary
        })
    
    # ì„±ë³„ í†µê³„ ê³„ì‚°
    male_salaries = [emp["salary"] for emp in employees if emp["gender"] == "ë‚¨"]
    female_salaries = [emp["salary"] for emp in employees if emp["gender"] == "ì—¬"]
    
    # ê²°ê³¼ ì¶œë ¥
    print("\n" + "="*60)
    print("ğŸ“ˆ ì„±ë³„ ì„ê¸ˆ ë¶„ì„ ê²°ê³¼")
    print("="*60)
    
    if male_salaries:
        male_avg = sum(male_salaries) / len(male_salaries)
        male_min = min(male_salaries)
        male_max = max(male_salaries)
        print(f"ğŸ‘¨ ë‚¨ì„± ì§ì› ({len(male_salaries)}ëª…)")
        print(f"   í‰ê·  ì—°ë´‰: {male_avg:,.0f}ë§Œì›")
        print(f"   ìµœì € ì—°ë´‰: {male_min:,}ë§Œì›")
        print(f"   ìµœê³  ì—°ë´‰: {male_max:,}ë§Œì›")
    else:
        print("ğŸ‘¨ ë‚¨ì„± ì§ì›: ë°ì´í„° ì—†ìŒ")
        male_avg = 0
    
    if female_salaries:
        female_avg = sum(female_salaries) / len(female_salaries)
        female_min = min(female_salaries)
        female_max = max(female_salaries)
        print(f"\nğŸ‘© ì—¬ì„± ì§ì› ({len(female_salaries)}ëª…)")
        print(f"   í‰ê·  ì—°ë´‰: {female_avg:,.0f}ë§Œì›")
        print(f"   ìµœì € ì—°ë´‰: {female_min:,}ë§Œì›")
        print(f"   ìµœê³  ì—°ë´‰: {female_max:,}ë§Œì›")
    else:
        print("\nğŸ‘© ì—¬ì„± ì§ì›: ë°ì´í„° ì—†ìŒ")
        female_avg = 0
    
    # ì„±ë³„ ê²©ì°¨ ë¶„ì„
    if male_salaries and female_salaries:
        gap = male_avg - female_avg
        gap_percentage = (gap / male_avg) * 100 if male_avg > 0 else 0
        
        print(f"\nâš–ï¸ ì„±ë³„ ì„ê¸ˆ ê²©ì°¨ ë¶„ì„")
        print(f"   ê²©ì°¨: {abs(gap):,.0f}ë§Œì›")
        print(f"   ê²©ì°¨ìœ¨: {abs(gap_percentage):.1f}%")
        
        if gap > 0:
            print(f"   âš ï¸ ë‚¨ì„± í‰ê· ì´ {gap:,.0f}ë§Œì› ë†’ìŒ")
        elif gap < 0:
            print(f"   âš ï¸ ì—¬ì„± í‰ê· ì´ {abs(gap):,.0f}ë§Œì› ë†’ìŒ")
        else:
            print(f"   âœ… ì„±ë³„ ì„ê¸ˆ ê²©ì°¨ ì—†ìŒ")
    
    # ì „ì²´ í†µê³„
    all_salaries = [emp["salary"] for emp in employees]
    if all_salaries:
        total_avg = sum(all_salaries) / len(all_salaries)
        print(f"\nğŸ“Š ì „ì²´ í‰ê·  ì—°ë´‰: {total_avg:,.0f}ë§Œì›")

# ì‹¤í–‰
# gender_salary_analysis()
```

```python
# ë¬¸ì œ 2: ëˆ„ì  í•©ê³„ ìˆ˜ì‹ ì¶œë ¥
# 1 â†’ 1
# 1 + 2 â†’ 3
# 1 + 2 + 3 â†’ 6
# ...
# 1 + 2 + ... + 10 â†’ 55

print("ğŸ“ ë¬¸ì œ 2: ëˆ„ì  í•©ê³„ ìˆ˜ì‹ ì¶œë ¥")
print("="*40)

# ë°©ë²• 1: ê¸°ë³¸ì ì¸ ë¬¸ìì—´ ì¡°ì‘
print("ë°©ë²• 1: ê¸°ë³¸ êµ¬í˜„")
for i in range(1, 11):
    total = 0
    formula = ""
    
    for j in range(1, i + 1):
        total += j
        if j == 1:
            formula = "1"
        else:
            formula += f" + {j}"
    
    print(f"{formula} â†’ {total}")

print("\në°©ë²• 2: join í•¨ìˆ˜ í™œìš©")
for i in range(1, 11):
    numbers = range(1, i + 1)
    formula = " + ".join(map(str, numbers))
    total = sum(numbers)
    print(f"{formula} â†’ {total}")

print("\në°©ë²• 3: ìˆ˜í•™ ê³µì‹ í™œìš© (Î£ = n(n+1)/2)")
for i in range(1, 11):
    formula = " + ".join(str(j) for j in range(1, i + 1))
    total = i * (i + 1) // 2  # ê°€ìš°ìŠ¤ ê³µì‹
    print(f"{formula} â†’ {total}")
```

```python
# ì˜ˆì œ 2: ê°’ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
def sigma(limit):
    """
    1ë¶€í„° limitê¹Œì§€ì˜ í•©ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜
    
    Args:
        limit (int): í•©ê³„ë¥¼ êµ¬í•  ë§ˆì§€ë§‰ ìˆ˜
    
    Returns:
        int: 1ë¶€í„° limitê¹Œì§€ì˜ í•©
    """
    total = 0
    for i in range(1, limit + 1):
        total += i
    return total

print("\nì˜ˆì œ 2: í•©ê³„ ê³„ì‚° í•¨ìˆ˜")
print("-" * 30)

# í•¨ìˆ˜ í™œìš©
test_cases = [10, 100, 1000]
for case in test_cases:
    result = sigma(case)
    # ê°€ìš°ìŠ¤ ê³µì‹ìœ¼ë¡œ ê²€ì¦
    expected = case * (case + 1) // 2
    print(f"sigma({case:4d}) = {result:8d} (ê²€ì¦: {result == expected})")

print(f"\nğŸš€ ì„±ëŠ¥ í…ŒìŠ¤íŠ¸: sigma(10000) = {sigma(10000):,}")
```

```python
# ì˜ˆì œ 3: ë‹¤ì–‘í•œ ë§¤ê°œë³€ìˆ˜ë¥¼ ê°€ì§„ í•¨ìˆ˜ë“¤
def calculate_grade(korean, english, math):
    """
    ì„¸ ê³¼ëª© ì ìˆ˜ë¡œ ì´ì , í‰ê· , ë“±ê¸‰ì„ ê³„ì‚°
    
    Returns:
        tuple: (ì´ì , í‰ê· , ë“±ê¸‰)
    """
    total = korean + english + math
    average = total / 3
    
    if average >= 90:
        grade = "ìˆ˜"
    elif average >= 80:
        grade = "ìš°"
    elif average >= 70:
        grade = "ë¯¸"
    elif average >= 60:
        grade = "ì–‘"
    else:
        grade = "ê°€"
    
    return total, average, grade

def format_currency(amount):
    """ê¸ˆì•¡ì„ ì²œ ë‹¨ìœ„ ì‰¼í‘œë¡œ í¬ë§·íŒ…"""
    return f"{amount:,}ì›"

def is_even(number):
    """ì§ìˆ˜ì¸ì§€ íŒë³„"""
    return number % 2 == 0

print("\nì˜ˆì œ 3: ë‹¤ì–‘í•œ í•¨ìˆ˜ í™œìš©")
print("="*40)

# ì„±ì  ê³„ì‚°
total, avg, grade = calculate_grade(85, 90, 88)
print(f"ğŸ“Š ì„±ì : ì´ì  {total}, í‰ê·  {avg:.1f}, ë“±ê¸‰ {grade}")

# ê¸ˆì•¡ í¬ë§·íŒ…
salary = 45000000
print(f"ğŸ’° ì—°ë´‰: {format_currency(salary)}")

# ì§ìˆ˜ íŒë³„
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = [n for n in numbers if is_even(n)]
print(f"ğŸ”¢ ì§ìˆ˜: {even_numbers}")

print("\nâœ… í•¨ìˆ˜ë¥¼ í™œìš©í•˜ë©´ ì½”ë“œê°€ ê¹”ë”í•˜ê³  ì¬ì‚¬ìš© ê°€ëŠ¥!")
```
