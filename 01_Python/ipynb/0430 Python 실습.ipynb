```python
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
```

## ğŸ” ì„ íƒ ì •ë ¬ (Selection Sort)

* ì´ **Nê°œì˜ ì›ì†Œ**ê°€ ìˆìœ¼ë©´, **N-1ë²ˆ ë°˜ë³µ**í•¨
* ê° ë°˜ë³µì—ì„œ \*\*ìµœì†Ÿê°’(ë˜ëŠ” ìµœëŒ“ê°’)\*\*ì˜ ìœ„ì¹˜ë¥¼ ì°¾ì•„ í˜„ì¬ ìœ„ì¹˜ì™€ **êµí™˜(swap)**

### ğŸ§ª Python êµ¬í˜„ ì˜ˆì‹œ

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

data = [5, 1, 2, 4, 3]
```

## ğŸ§  ìš”ì•½

* ì„ íƒ ì •ë ¬ì€ **ë‹¨ìˆœí•˜ê³  êµ¬í˜„í•˜ê¸° ì‰¬ì›€**
* ê·¸ëŸ¬ë‚˜ \*\*ì‹œê°„ë³µì¡ë„ O(nÂ²)\*\*ë¡œ ëŒ€ê·œëª¨ ë°ì´í„°ì—ëŠ” ë¶€ì í•©
* í•™ìŠµìš©/ì´í•´ìš©ìœ¼ë¡œ ì í•©í•˜ê³ , ì •ë ¬ ì•Œê³ ë¦¬ì¦˜ì˜ ê¸°ë³¸ ê°œë… íŒŒì•…ì— ìœ ìš©


````python
def selectSort(dataList, key):
    """
    ì„ íƒ ì •ë ¬(Selection Sort) ì•Œê³ ë¦¬ì¦˜.
    dataListì˜ ì›ì†Œë¥¼ key í•¨ìˆ˜ ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬í•œ ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜.
    ì›ë³¸ dataListëŠ” ë³€ê²½í•˜ì§€ ì•ŠìŒ(ê¹Šì€ ë³µì‚¬ ì‚¬ìš©).

    Args:
        dataList (list): ì •ë ¬í•  ë°ì´í„° ë¦¬ìŠ¤íŠ¸
        key (function): ì •ë ¬ ê¸°ì¤€ì´ ë  í•¨ìˆ˜

    Returns:
        list: ì •ë ¬ëœ ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸
    """
    # ê¹Šì€ ë³µì‚¬(ì›ë³¸ ë³€ê²½ ë°©ì§€)
    aList = [x for x in dataList]
    n = len(aList)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            # key í•¨ìˆ˜ ê¸°ì¤€ìœ¼ë¡œ ë” ì‘ì€ ê°’ ì°¾ê¸°
            if key(aList[j]) < key(aList[min_idx]):
                min_idx = j
        # ìµœì†Ÿê°’ì„ í˜„ì¬ ìœ„ì¹˜ì™€ êµí™˜
        aList[i], aList[min_idx] = aList[min_idx], aList[i]
    return aList
````

````python
# ìˆ«ì ë¦¬ìŠ¤íŠ¸ ì˜ˆì‹œ
a = [5, 1, 2, 4, 3]
b = selectSort(a, key=lambda x: x)
print(a)  # ì›ë³¸ ë¦¬ìŠ¤íŠ¸
print(b)  # ì •ë ¬ëœ ë¦¬ìŠ¤íŠ¸
````
