# Part 1: ÌôòÍ≤Ω ÏÑ§Ï†ï & Í∏∞Ï¥à

## 1.1 ÎùºÏù¥Î∏åÎü¨Î¶¨ Import Î∞è ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï

Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌïú ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨Îì§ÏùÑ importÌïòÍ≥†, ÌïúÍ∏Ä ÌÖçÏä§Ìä∏Í∞Ä Ï†úÎåÄÎ°ú ÌëúÏãúÎêòÎèÑÎ°ù Ìè∞Ìä∏Î•º ÏÑ§Ï†ïÌï©ÎãàÎã§.

**ÌïµÏã¨ Í∞úÎÖê**:
- `matplotlib.pyplot`: Í∏∞Î≥∏ Ï∞®Ìä∏ ÏÉùÏÑ±
- `font_manager`: ÏãúÏä§ÌÖú Ìè∞Ìä∏ Í¥ÄÎ¶¨
- `rcParams`: matplotlib Ï†ÑÏó≠ ÏÑ§Ï†ï

```python
# 1.1.1 ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ Import
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from matplotlib import font_manager, rc
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ ÎùºÏù¥Î∏åÎü¨Î¶¨ Import ÏôÑÎ£å")

# 1.1.2 ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
try:
    # Windows ÏãúÏä§ÌÖú Ìè∞Ìä∏ ÏÇ¨Ïö© (H2GTRM - Ìú¥Î®ºÎ™®ÏùåT)
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/H2GTRM.TTF").get_name()
    print(f"‚úÖ ÏÇ¨Ïö©Ìï† Ìè∞Ìä∏: {font_name}")
except:
    # Ìè∞Ìä∏Î•º Ï∞æÏùÑ Ïàò ÏóÜÎäî Í≤ΩÏö∞ Í∏∞Î≥∏ Ìè∞Ìä∏ ÏÇ¨Ïö©
    font_name = "DejaVu Sans"
    print(f"‚ö†Ô∏è  Í∏∞Î≥∏ Ìè∞Ìä∏ ÏÇ¨Ïö©: {font_name}")

# 1.1.3 matplotlib Ï†ÑÏó≠ ÏÑ§Ï†ï
plt.rcParams['font.family'] = font_name          # Ìè∞Ìä∏ Ìå®Î∞ÄÎ¶¨ ÏÑ§Ï†ï
plt.rcParams['axes.unicode_minus'] = False       # ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
plt.rcParams['figure.figsize'] = (10, 6)         # Í∏∞Î≥∏ Ï∞®Ìä∏ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
plt.rcParams['axes.grid'] = True                 # Í≤©Ïûê ÌëúÏãú
plt.rcParams['grid.alpha'] = 0.3                 # Í≤©Ïûê Ìà¨Î™ÖÎèÑ

print("‚úÖ matplotlib ÌôòÍ≤Ω ÏÑ§Ï†ï ÏôÑÎ£å")
```

```python
# Part 2: Í∏∞Î≥∏ Ï∞®Ìä∏ ÏÉùÏÑ±

## 2.1 ÎùºÏù∏ Ï∞®Ìä∏ (Line Chart) Í∏∞Ï¥à

# 2.1.1 ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
y1 = [1, 2, 3, 4, 5]  # yÏ∂ï Îç∞Ïù¥ÌÑ∞ (list ÌÉÄÏûÖ)
x1 = [1, 2, 3, 4, 5]  # xÏ∂ï Îç∞Ïù¥ÌÑ∞ (list ÌÉÄÏûÖ)

# 2.1.2 List vs NumPy Î≤°ÌÑ∞ Ïó∞ÏÇ∞Ïùò Ï∞®Ïù¥Ï†ê Ïù¥Ìï¥
print("üìå ListÏôÄ NumPy Î≤°ÌÑ∞ Ïó∞ÏÇ∞ ÎπÑÍµê:")
print(f"ÏõêÎ≥∏ x1: {x1}")

# ListÏóêÏÑú *2 Ïó∞ÏÇ∞ ‚Üí ÏöîÏÜåÍ∞Ä 2Î≤à Î∞òÎ≥µÎê®
x1_list_mult = x1 * 2
print(f"List * 2: {x1_list_mult}")

# List comprehensionÏúºÎ°ú Í∞Å ÏöîÏÜåÏóê 2Î•º Í≥±Ìï®
x2 = [n * 2 for n in x1] 
print(f"List comprehension [n*2 for n in x1]: {x2}")

# NumPy arrayÏóêÏÑú *2 Ïó∞ÏÇ∞ ‚Üí Í∞Å ÏöîÏÜåÏóê 2Í∞Ä Í≥±Ìï¥Ïßê
x1_numpy = np.array(x1)
x1_numpy_mult = x1_numpy * 2
print(f"NumPy array * 2: {x1_numpy_mult}")

# 2.1.3 Í∏∞Î≥∏ ÎùºÏù∏ Ï∞®Ìä∏ ÏÉùÏÑ±
plt.figure(figsize=(12, 8))
plt.plot(x1, y1, label='ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞', marker='o', linewidth=2, markersize=8)
plt.plot(x2, y1, label='xÏ∂ï 2Î∞∞ ÌôïÏû•', marker='s', linewidth=2, markersize=8)

plt.title('ÎùºÏù∏ Ï∞®Ìä∏ Í∏∞Î≥∏ ÏòàÏ†ú', fontsize=16, fontweight='bold')
plt.xlabel('XÏ∂ï (ÎèÖÎ¶ΩÎ≥ÄÏàò)', fontsize=12)
plt.ylabel('YÏ∂ï (Ï¢ÖÏÜçÎ≥ÄÏàò)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# Ï∞®Ìä∏ ÎØ∏ÏÑ∏ Ï°∞Ï†ï
plt.tight_layout()
plt.show()

print("‚úÖ Í∏∞Î≥∏ ÎùºÏù∏ Ï∞®Ìä∏ ÏôÑÏÑ±")
```

# Part 3: Í≥†Í∏â ÏãúÍ∞ÅÌôî Í∏∞Î≤ï

## 3.1 Seaborn Ïä§ÌÉÄÏùº Ï†ÅÏö©

SeabornÏùÄ matplotlibÏùÑ Í∏∞Î∞òÏúºÎ°ú Ìïú Í≥†Í∏â ÌÜµÍ≥Ñ ÏãúÍ∞ÅÌôî ÎùºÏù¥Î∏åÎü¨Î¶¨ÏûÖÎãàÎã§. Îçî ÏïÑÎ¶ÑÎãµÍ≥† Ï†ÑÎ¨∏Ï†ÅÏù∏ Ï∞®Ìä∏Î•º ÏâΩÍ≤å ÎßåÎì§ Ïàò ÏûàÏäµÎãàÎã§.

**Ï£ºÏöî Seaborn Ïä§ÌÉÄÏùº**:
- `darkgrid`: Ïñ¥ÎëêÏö¥ Î∞∞Í≤Ω + Í≤©Ïûê
- `whitegrid`: Î∞ùÏùÄ Î∞∞Í≤Ω + Í≤©Ïûê  
- `dark`: Ïñ¥ÎëêÏö¥ Î∞∞Í≤Ω (Í≤©Ïûê ÏóÜÏùå)
- `white`: Î∞ùÏùÄ Î∞∞Í≤Ω (Í≤©Ïûê ÏóÜÏùå)
- `ticks`: ÏµúÏÜåÌïúÏùò Ïä§ÌÉÄÏùº

```python
# 3.1.1 Seaborn Import Î∞è Ïä§ÌÉÄÏùº ÏÑ§Ï†ï
import seaborn as sns 

# Ï£ºÏùò: seaborn Ïä§ÌÉÄÏùºÏùÄ Ìè∞Ìä∏ ÏÑ§Ï†ïÎ≥¥Îã§ Î®ºÏ†Ä Ï†ÅÏö©Ìï¥Ïïº Ìï®
sns.set_style('whitegrid')  # ÍπîÎÅîÌïú Ìù∞ÏÉâ Î∞∞Í≤Ω + Í≤©Ïûê Ïä§ÌÉÄÏùº

# 3.1.2 Ìè∞Ìä∏ ÏÑ§Ï†ï Ïû¨Ï†ÅÏö© (seaborn Ï†ÅÏö© ÌõÑ ÌïÑÏàò)
plt.rcParams['font.family'] = font_name     
plt.rcParams['axes.unicode_minus'] = False 

print("‚úÖ Seaborn Ïä§ÌÉÄÏùº Ï†ÅÏö© ÏôÑÎ£å")

# 3.1.3 Ïó∞ÏÜç Ìï®Ïàò Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
# np.linspace(): ÏßÄÏ†ïÎêú Íµ¨Í∞ÑÏùÑ Í∑†Îì±ÌïòÍ≤å ÎÇòÎàÑÎäî Ìï®Ïàò
x = np.linspace(0, 2, 100)  # 0~2 Íµ¨Í∞ÑÏùÑ 100Í∞úÎ°ú Î∂ÑÌï†
print(f"üìä ÏÉùÏÑ±Îêú x Îç∞Ïù¥ÌÑ∞ Ï†ïÎ≥¥:")
print(f"   - Î≤îÏúÑ: {x.min():.3f} ~ {x.max():.3f}")
print(f"   - Îç∞Ïù¥ÌÑ∞ Í∞úÏàò: {len(x)}Í∞ú")
print(f"   - Ï≤òÏùå 10Í∞ú Í∞í: {x[:10]}")

# Í∞Å Ìï®ÏàòÎ≥Ñ yÍ∞í Í≥ÑÏÇ∞ ÎØ∏Î¶¨Î≥¥Í∏∞
y_linear = x
y_quadratic = x**2  
y_cubic = x**3

print(f"\nüìà x=2Ïùº Îïå Ìï®ÏàòÍ∞í ÎπÑÍµê:")
print(f"   - ÏÑ†Ìòï Ìï®Ïàò (y=x): {y_linear[-1]:.3f}")
print(f"   - 2Ï∞® Ìï®Ïàò (y=x¬≤): {y_quadratic[-1]:.3f}")  
print(f"   - 3Ï∞® Ìï®Ïàò (y=x¬≥): {y_cubic[-1]:.3f}")
```

```python
# 3.2 Îã§Ï§ë Ìï®Ïàò ÎπÑÍµê Ï∞®Ìä∏

# 3.2.1 Í≥†Í∏â Ï∞®Ìä∏ ÏÉùÏÑ±
plt.figure(figsize=(14, 10))

# Î©îÏù∏ Ï∞®Ìä∏
plt.subplot(2, 1, 1)
plt.plot(x, x,    label='ÏÑ†Ìòï Ìï®Ïàò (y = x)',   color='#2E8B57', linewidth=3, alpha=0.8)
plt.plot(x, x**2, label='2Ï∞® Ìï®Ïàò (y = x¬≤)', color='#4169E1', linewidth=3, alpha=0.8)  
plt.plot(x, x**3, label='3Ï∞® Ìï®Ïàò (y = x¬≥)', color='#DC143C', linewidth=3, alpha=0.8)

plt.title('Îã§Ìï≠Ìï®Ïàò ÏÑ±Ïû• Ìå®ÌÑ¥ ÎπÑÍµê Î∂ÑÏÑù', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('XÍ∞í (ÏûÖÎ†•)', fontsize=14)
plt.ylabel('YÍ∞í (Ï∂úÎ†•)', fontsize=14)
plt.legend(fontsize=12, loc='upper left')
plt.grid(True, alpha=0.3)

# ÌäπÏ†ï ÏßÄÏ†êÏóêÏÑúÏùò Ìï®ÏàòÍ∞í ÌëúÏãú
highlight_x = 1.5
plt.axvline(x=highlight_x, color='gray', linestyle='--', alpha=0.7)
plt.plot(highlight_x, highlight_x, 'o', color='#2E8B57', markersize=10)
plt.plot(highlight_x, highlight_x**2, 'o', color='#4169E1', markersize=10)
plt.plot(highlight_x, highlight_x**3, 'o', color='#DC143C', markersize=10)

# 3.2.2 Ìï®ÏàòÎ≥Ñ Ï¶ùÍ∞ÄÏú® ÎπÑÍµê (ÏÑúÎ∏åÏ∞®Ìä∏)
plt.subplot(2, 1, 2)
# 1Ï∞® ÎèÑÌï®Ïàò (Í∏∞Ïö∏Í∏∞) Í≥ÑÏÇ∞
dx = x[1] - x[0]  # x Í∞ÑÍ≤©
dy_linear = np.gradient(x, dx)
dy_quadratic = np.gradient(x**2, dx)  
dy_cubic = np.gradient(x**3, dx)

plt.plot(x, dy_linear, label='ÏÑ†ÌòïÌï®Ïàò Í∏∞Ïö∏Í∏∞', color='#2E8B57', linewidth=2)
plt.plot(x, dy_quadratic, label='2Ï∞®Ìï®Ïàò Í∏∞Ïö∏Í∏∞', color='#4169E1', linewidth=2)
plt.plot(x, dy_cubic, label='3Ï∞®Ìï®Ïàò Í∏∞Ïö∏Í∏∞', color='#DC143C', linewidth=2)

plt.title('Ìï®ÏàòÎ≥Ñ Î≥ÄÌôîÏú®(Í∏∞Ïö∏Í∏∞) ÎπÑÍµê', fontsize=14, fontweight='bold')
plt.xlabel('XÍ∞í', fontsize=12)
plt.ylabel('Î≥ÄÌôîÏú® (dy/dx)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3.2.3 ÏàòÏπòÏ†Å Î∂ÑÏÑù Í≤∞Í≥º
print("üìä x=1.5ÏóêÏÑúÏùò Ìï®ÏàòÍ∞í Î∂ÑÏÑù:")
print(f"   - ÏÑ†Ìòï Ìï®Ïàò: {highlight_x:.1f}")
print(f"   - 2Ï∞® Ìï®Ïàò: {highlight_x**2:.3f}")
print(f"   - 3Ï∞® Ìï®Ïàò: {highlight_x**3:.3f}")
print(f"\nüí° Ìï¥ÏÑù: xÍ∞íÏù¥ Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù Í≥†Ï∞® Ìï®ÏàòÏùò Ï¶ùÍ∞ÄÌè≠Ïù¥ Í∏âÍ≤©Ìûà Ïª§Ïßê")
```

# Part 4: ÌÜµÍ≥ÑÏ†Å ÏãúÍ∞ÅÌôî

## 4.1 ÌôïÎ•†Î∂ÑÌè¨ÏôÄ ÌûàÏä§ÌÜ†Í∑∏Îû®

ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨Î•º ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÌååÏïÖÌïòÎäî Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù¥Í≥† Ï§ëÏöîÌïú ÎèÑÍµ¨ÏûÖÎãàÎã§.

**ÌïµÏã¨ Í∞úÎÖê**:
- **Ï†ïÍ∑úÎ∂ÑÌè¨**: ÏûêÏó∞Í≥ÑÏóêÏÑú Í∞ÄÏû• ÌùîÌûà ÎÇòÌÉÄÎÇòÎäî Ï¢Ö Î™®Ïñë Î∂ÑÌè¨
- **ÌûàÏä§ÌÜ†Í∑∏Îû®**: Îç∞Ïù¥ÌÑ∞Î•º Íµ¨Í∞ÑÎ≥ÑÎ°ú ÎÇòÎàÑÏñ¥ ÎπàÎèÑÎ•º ÌëúÏãú
- **Ïª§ÎÑê Î∞ÄÎèÑ Ï∂îÏ†ï(KDE)**: ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Î∂ÄÎìúÎü¨Ïö¥ Í≥°ÏÑ†ÏúºÎ°ú Í∑ºÏÇ¨

**Ï†ïÍ∑úÎ∂ÑÌè¨ Îß§Í∞úÎ≥ÄÏàò**:
- `loc`: ÌèâÍ∑† (Î∂ÑÌè¨Ïùò Ï§ëÏã¨)
- `scale`: ÌëúÏ§ÄÌé∏Ï∞® (Î∂ÑÌè¨Ïùò ÌçºÏßê Ï†ïÎèÑ)  
- `size`: ÏÉùÏÑ±Ìï† Îç∞Ïù¥ÌÑ∞ Í∞úÏàò

```python
# 4.1.1 Ï†ïÍ∑úÎ∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±

# ÏãúÎìú ÏÑ§Ï†ï (Ïû¨ÌòÑ Í∞ÄÎä•Ìïú Í≤∞Í≥ºÎ•º ÏúÑÌï¥)
np.random.seed(42)

# Case 1: ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ï†ïÍ∑úÎ∂ÑÌè¨ (ÌèâÍ∑†=70, ÌëúÏ§ÄÌé∏Ï∞®=20)
# Ïòà: ÌïôÏÉùÎì§Ïùò ÏãúÌóò Ï†êÏàò Î∂ÑÌè¨
data1 = np.random.normal(loc=70, scale=20, size=500)

# Case 2: ÌëúÏ§Ä Ï†ïÍ∑úÎ∂ÑÌè¨ (ÌèâÍ∑†=0, ÌëúÏ§ÄÌé∏Ï∞®=1)  
# ÌÜµÍ≥ÑÌïôÏóêÏÑú Í∞ÄÏû• Í∏∞Î≥∏Ïù¥ ÎêòÎäî Î∂ÑÌè¨
data2 = np.random.normal(loc=0, scale=1, size=1000)

# 4.1.2 ÏÉùÏÑ±Îêú Îç∞Ïù¥ÌÑ∞Ïùò ÌÜµÍ≥ÑÏ†Å ÌäπÏÑ± Î∂ÑÏÑù
print("üìä ÏÉùÏÑ±Îêú Îç∞Ïù¥ÌÑ∞Ïùò ÌÜµÍ≥Ñ Î∂ÑÏÑù")
print("=" * 50)

print(f"üìà Case 1 - ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ï†ïÍ∑úÎ∂ÑÌè¨ (Î™©Ìëú: ÌèâÍ∑†=70, ÌëúÏ§ÄÌé∏Ï∞®=20)")
print(f"   - Ïã§Ï†ú ÌèâÍ∑†: {data1.mean():.2f}")
print(f"   - Ïã§Ï†ú ÌëúÏ§ÄÌé∏Ï∞®: {data1.std():.2f}")
print(f"   - ÏµúÏÜüÍ∞í: {data1.min():.2f}")
print(f"   - ÏµúÎåìÍ∞í: {data1.max():.2f}")
print(f"   - Ï§ëÏïôÍ∞í: {np.median(data1):.2f}")

print(f"\nüìà Case 2 - ÌëúÏ§Ä Ï†ïÍ∑úÎ∂ÑÌè¨ (Î™©Ìëú: ÌèâÍ∑†=0, ÌëúÏ§ÄÌé∏Ï∞®=1)")
print(f"   - Ïã§Ï†ú ÌèâÍ∑†: {data2.mean():.2f}")
print(f"   - Ïã§Ï†ú ÌëúÏ§ÄÌé∏Ï∞®: {data2.std():.2f}")
print(f"   - ÏµúÏÜüÍ∞í: {data2.min():.2f}")
print(f"   - ÏµúÎåìÍ∞í: {data2.max():.2f}")
print(f"   - Ï§ëÏïôÍ∞í: {np.median(data2):.2f}")

# 4.1.3 68-95-99.7 Í∑úÏπô Í≤ÄÏ¶ù (ÌëúÏ§Ä Ï†ïÍ∑úÎ∂ÑÌè¨)
within_1std = np.sum(np.abs(data2) <= 1) / len(data2) * 100
within_2std = np.sum(np.abs(data2) <= 2) / len(data2) * 100
within_3std = np.sum(np.abs(data2) <= 3) / len(data2) * 100

print(f"\nüéØ Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò 68-95-99.7 Í∑úÏπô Í≤ÄÏ¶ù:")
print(f"   - 1 ÌëúÏ§ÄÌé∏Ï∞® ÎÇ¥: {within_1std:.1f}% (Ïù¥Î°†Í∞í: 68.3%)")
print(f"   - 2 ÌëúÏ§ÄÌé∏Ï∞® ÎÇ¥: {within_2std:.1f}% (Ïù¥Î°†Í∞í: 95.4%)")  
print(f"   - 3 ÌëúÏ§ÄÌé∏Ï∞® ÎÇ¥: {within_3std:.1f}% (Ïù¥Î°†Í∞í: 99.7%)")
```

# Part 5: Ïã§Î¨¥ ÏùëÏö© - Îã§ÏñëÌïú Î∂ÑÌè¨ ÎπÑÍµê Î∂ÑÏÑù

## 5.1 Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÎßåÎÇ† Ïàò ÏûàÎäî Îã§ÏñëÌïú Î∂ÑÌè¨Îì§

- **ÏßÄÏàòÎ∂ÑÌè¨**: ÎåÄÍ∏∞ÏãúÍ∞Ñ, ÏÉùÏ°¥ Î∂ÑÏÑù, Í≥†Ïû•Î•† Îì±
- **Î≤†ÌÉÄÎ∂ÑÌè¨**: ÌôïÎ•†Ïùò ÌôïÎ•†, ÎπÑÏú® Îç∞Ïù¥ÌÑ∞ Îì±

**Ïã§Î¨¥ ÌôúÏö© ÏòàÏãú**:
- A/B ÌÖåÏä§Ìä∏ Í≤∞Í≥º Î∂ÑÏÑù
- ÌíàÏßà Í¥ÄÎ¶¨ Î∞è Í≥µÏ†ï Î∂ÑÏÑù  
- Í≥†Í∞ù ÌñâÎèô Ìå®ÌÑ¥ Î∂ÑÏÑù
- Í∏àÏúµ Î¶¨Ïä§ÌÅ¨ Î™®Îç∏ÎßÅ

```python
# 5.1.1 Ïã§Î¨¥ ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±

# ÏãúÎìú Í≥†Ï†ï (Ïû¨ÌòÑ Í∞ÄÎä•Ìïú Í≤∞Í≥º)
np.random.seed(123)

print("üéØ Ïã§Î¨¥ ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±")
print("=" * 50)

# ÏãúÎÇòÎ¶¨Ïò§ 1: Í≥†Í∞ù ÎßåÏ°±ÎèÑ Ï†êÏàò (Ï†ïÍ∑úÎ∂ÑÌè¨)
# 1-10Ï†ê Ï≤ôÎèÑ, ÌèâÍ∑† 7Ï†ê, ÌëúÏ§ÄÌé∏Ï∞® 1.5Ï†ê
satisfaction_scores = np.random.normal(loc=7, scale=1.5, size=1000)
satisfaction_scores = np.clip(satisfaction_scores, 1, 10)  # 1-10Ï†ê Î≤îÏúÑÎ°ú Ï†úÌïú

# ÏãúÎÇòÎ¶¨Ïò§ 2: ÏõπÏÇ¨Ïù¥Ìä∏ Î°úÎî© ÏãúÍ∞Ñ (ÏßÄÏàòÎ∂ÑÌè¨)  
# ÌèâÍ∑† 2Ï¥à, ÎåÄÎ∂ÄÎ∂Ñ Îπ†Î•¥ÏßÄÎßå Í∞ÄÎÅî Îß§Ïö∞ ÎäêÎ¶º
loading_times = np.random.exponential(scale=2, size=1000)

# ÏãúÎÇòÎ¶¨Ïò§ 3: Ï£ºÏÇ¨ÏúÑ Í≤∞Í≥º ÏãúÎÆ¨Î†àÏù¥ÏÖò (Ïù¥ÏÇ∞ Í∑†Îì±Î∂ÑÌè¨)
# 1-6ÍπåÏßÄ ÎèôÏùºÌïú ÌôïÎ•†
dice_results = np.random.randint(1, 7, size=1000)

# ÏãúÎÇòÎ¶¨Ïò§ 4: ÌîÑÎ°úÏ†ùÌä∏ ÏôÑÎ£åÏú® (Î≤†ÌÉÄÎ∂ÑÌè¨)
# Œ±=2, Œ≤=5 ‚Üí ÎÇÆÏùÄ ÏôÑÎ£åÏú®Ïóê Ìé∏Ï§ë
completion_rates = np.random.beta(a=2, b=5, size=1000) * 100

# 5.1.2 ÏÉùÏÑ±Îêú Îç∞Ïù¥ÌÑ∞ ÌÜµÍ≥Ñ ÏöîÏïΩ
scenarios = [
    ("Í≥†Í∞ù ÎßåÏ°±ÎèÑ", satisfaction_scores, "Ï†ê"),
    ("Î°úÎî© ÏãúÍ∞Ñ", loading_times, "Ï¥à"), 
    ("Ï£ºÏÇ¨ÏúÑ Í≤∞Í≥º", dice_results, ""),
    ("ÌîÑÎ°úÏ†ùÌä∏ ÏôÑÎ£åÏú®", completion_rates, "%")
]

for name, data, unit in scenarios:
    print(f"\nüìä {name} ÌÜµÍ≥Ñ:")
    print(f"   - ÌèâÍ∑†: {np.mean(data):.2f}{unit}")
    print(f"   - Ï§ëÏïôÍ∞í: {np.median(data):.2f}{unit}")
    print(f"   - ÌëúÏ§ÄÌé∏Ï∞®: {np.std(data):.2f}{unit}")
    print(f"   - Î≤îÏúÑ: {np.min(data):.2f} ~ {np.max(data):.2f}{unit}")

# 5.1.3 Ï¢ÖÌï© ÏãúÍ∞ÅÌôî
fig = plt.figure(figsize=(20, 15))
gs = fig.add_gridspec(3, 4, hspace=0.3, wspace=0.3)

# Í∞Å ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ ÌûàÏä§ÌÜ†Í∑∏Îû®
scenarios_data = [satisfaction_scores, loading_times, dice_results, completion_rates]
scenarios_names = ["Í≥†Í∞ù ÎßåÏ°±ÎèÑ (Ï†ïÍ∑úÎ∂ÑÌè¨)", "Î°úÎî© ÏãúÍ∞Ñ (ÏßÄÏàòÎ∂ÑÌè¨)", "Ï£ºÏÇ¨ÏúÑ Í≤∞Í≥º (Í∑†Îì±Î∂ÑÌè¨)", "ÌîÑÎ°úÏ†ùÌä∏ ÏôÑÎ£åÏú® (Î≤†ÌÉÄÎ∂ÑÌè¨)"]
colors = ['skyblue', 'lightgreen', 'coral', 'plum']

for i, (data, name, color) in enumerate(zip(scenarios_data, scenarios_names, colors)):
    ax = fig.add_subplot(gs[0, i])
    
    if i == 2:  # Ï£ºÏÇ¨ÏúÑÎäî Ïù¥ÏÇ∞Ìòï Îç∞Ïù¥ÌÑ∞
        bins = np.arange(0.5, 7.5, 1)
        ax.hist(data, bins=bins, alpha=0.7, color=color, edgecolor='black')
    else:
        ax.hist(data, bins=25, alpha=0.7, color=color, edgecolor='black', density=True)
        if i != 2:  # Ï£ºÏÇ¨ÏúÑÍ∞Ä ÏïÑÎãå Í≤ΩÏö∞ KDE Ï∂îÍ∞Ä
            from scipy import stats
            x_range = np.linspace(data.min(), data.max(), 100)
            kde = stats.gaussian_kde(data)
            ax.plot(x_range, kde(x_range), 'r-', linewidth=2, alpha=0.8)
    
    ax.set_title(name, fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3)

# Î∞ïÏä§ÌîåÎ°Ø ÎπÑÍµê (ÏÉÅÎã® Ï§ëÏïô)
ax_box = fig.add_subplot(gs[1, :2])
box_data = [satisfaction_scores, loading_times/loading_times.max()*10, completion_rates/10]  # Ïä§ÏºÄÏùº Ï°∞Ï†ï
box_labels = ['ÎßåÏ°±ÎèÑ\n(1-10)', 'Î°úÎî©ÏãúÍ∞Ñ\n(Ï†ïÍ∑úÌôî)', 'ÏôÑÎ£åÏú®\n(0-10)']
bp = ax_box.boxplot(box_data, labels=box_labels, patch_artist=True)
for patch, color in zip(bp['boxes'], ['skyblue', 'lightgreen', 'plum']):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
ax_box.set_title('Î∂ÑÌè¨ ÌòïÌÉú ÎπÑÍµê (Î∞ïÏä§ÌîåÎ°Ø)', fontsize=14, fontweight='bold')
ax_box.grid(True, alpha=0.3)

# Î∂ÑÌè¨Î≥Ñ ÌäπÏÑ± ÎπÑÍµê ÌÖåÏù¥Î∏î
ax_table = fig.add_subplot(gs[1, 2:])
ax_table.axis('off')

# ÌÖåÏù¥Î∏î Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
table_data = []
for name, data, unit in scenarios:
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    table_data.append([
        name,
        f"{np.mean(data):.1f}{unit}",
        f"{np.std(data):.1f}{unit}", 
        f"{skewness:.2f}",
        f"{kurtosis:.2f}"
    ])

table = ax_table.table(
    cellText=table_data,
    colLabels=['Î∂ÑÌè¨Î™Ö', 'ÌèâÍ∑†', 'ÌëúÏ§ÄÌé∏Ï∞®', 'ÏôúÎèÑ', 'Ï≤®ÎèÑ'],
    cellLoc='center',
    loc='center',
    bbox=[0, 0, 1, 1]
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 2)
ax_table.set_title('Î∂ÑÌè¨Î≥Ñ ÌÜµÍ≥Ñ ÏßÄÌëú ÎπÑÍµê', fontsize=14, fontweight='bold', pad=20)

# Q-Q ÌîåÎ°Ø (Ï†ïÍ∑úÏÑ± Í≤ÄÏ†ï)
ax_qq = fig.add_subplot(gs[2, :2])
stats.probplot(satisfaction_scores, dist="norm", plot=ax_qq)
ax_qq.set_title('Í≥†Í∞ù ÎßåÏ°±ÎèÑ Q-Q ÌîåÎ°Ø (Ï†ïÍ∑úÏÑ± Í≤ÄÏ†ï)', fontsize=12, fontweight='bold')
ax_qq.grid(True, alpha=0.3)

# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÏÇ∞Ï†êÎèÑ 
ax_scatter = fig.add_subplot(gs[2, 2:])
# Í∞ÄÏÉÅÏùò ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÏÉùÏÑ± (ÎßåÏ°±ÎèÑ vs Î°úÎî©ÏãúÍ∞Ñ)
scatter_loading = np.random.exponential(scale=2, size=len(satisfaction_scores))
ax_scatter.scatter(satisfaction_scores, scatter_loading, alpha=0.6, color='orange')
ax_scatter.set_xlabel('Í≥†Í∞ù ÎßåÏ°±ÎèÑ')
ax_scatter.set_ylabel('Î°úÎî© ÏãúÍ∞Ñ (Ï¥à)')
ax_scatter.set_title('ÎßåÏ°±ÎèÑ vs Î°úÎî©ÏãúÍ∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ', fontsize=12, fontweight='bold')
ax_scatter.grid(True, alpha=0.3)

# ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞ Î∞è ÌëúÏãú
correlation = np.corrcoef(satisfaction_scores, scatter_loading)[0, 1]
ax_scatter.text(0.05, 0.95, f'ÏÉÅÍ¥ÄÍ≥ÑÏàò: {correlation:.3f}', transform=ax_scatter.transAxes, 
                fontsize=11, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.suptitle('Ïã§Î¨¥ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨ Ï¢ÖÌï© Î∂ÑÏÑù ÎåÄÏãúÎ≥¥Îìú', fontsize=20, fontweight='bold', y=0.98)
plt.show()

print("\n‚úÖ Ïã§Î¨¥ ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ Î∂ÑÌè¨ Î∂ÑÏÑù ÏôÑÎ£å")
```
