```python
import pandas as pd

data1 = {'eng': 88, 'math': 90, 'kor': 85}
data2 = {'math': 85, 'kor': 92, 'eng': 90}
data3 = {'kor': 78, 'eng': 85, 'math': 92}
data4 = {'math': 88, 'kor': 95, 'eng': 92}

# í‚¤ê°€ ì—†ìœ¼ë©´ NaN ì¶œë ¥.  NaNì€ ëˆ„ë½ëœ ê°’ì„ ì˜ë¯¸
s1 = pd.Series(data1)
s2 = pd.Series(data2)
s3 = pd.Series(data3)
s4 = pd.Series(data4)

print("s1:\n", s1)
print("s2:\n", s2)
print("s3:\n", s3)
print("s4:\n", s4)
```

```python
result = s1+s2+s3+s4
print("result:\n", result)
```

```python
result2 = s1.add(s2, fill_value=0).add(s3, fill_value=0).add(s4, fill_value=0)
print("result2:\n", result2)
```

```python
# ë°ì´í„°í”„ë ˆì„ ìƒì„±
import pandas as pd

data = {
    'name': ['ê¹€ì² ìˆ˜', 'ì´ì˜í¬', 'ë°•ë¯¼ìˆ˜', 'ìµœì§€ì˜', 'ì •í˜„ìš°'],
    'kor': [85, 92, 78, 95, 88],
    'math': [90, 85, 92, 88, 95],
    'eng': [88, 90, 85, 92, 87]
}
df = pd.DataFrame(data)

# í–‰ ì¶”ê°€
df.loc[len(df)] = data1
df.loc[len(df)] = data2
df.loc[len(df)] = data3
df.loc[len(df)] = data4

# ì¸ë±ìŠ¤ ì¬ì„¤ì •(í•„ìš”ì‹œ)
df = df.reset_index(drop=True)

# ì›í•˜ëŠ” í–‰, ì—´ ì¶œë ¥
print(df.iloc[[1, 3, 5], [0, 1, 2]])
```

```python
df = pd.concat([
    pd.DataFrame([data1]),
    pd.DataFrame([data2]),
    pd.DataFrame([data3]),
    pd.DataFrame([data4]),
    ], ignore_index = True)
print("\n concat ê²°ê³¼ : ")
print(df)
```

```python
df['total'] = df['kor'] + df['eng'] + df['math']
df['avg'] = round(df['total'] / 3)

print(df)
```
